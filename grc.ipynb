Modify the `perform_grc_checks` function to accept a `recipe_name`. Based on this name, it will dynamically retrieve and apply the corresponding keywords and categorization rules from `ANALYSIS_RECIPES` to perform granular GRC checks and generate a tailored report.

This also necessitates updating `process_document_for_docling` to use recipe-specific keywords and `main_grc_doc_processor` to pass the recipe name through. I will regenerate the entire code cell `bdbff331` with these modifications, ensuring all related functions are updated for recipe-based analysis.



```python
import os
import PyPDF2 # For PDF files
import docx # For DOCX files

# Function to simulate document parsing/reading
def read_document(file_path):
    print(f"Reading document: {file_path}")
    try:
        if file_path.lower().endswith('.pdf'):
            with open(file_path, 'rb') as f:
                reader = PyPDF2.PdfReader(f)
                content = "".join([page.extract_text() for page in reader.pages if page.extract_text()])
            return content
        elif file_path.lower().endswith('.docx'):
            document = docx.Document(file_path)
            content = "\n".join([para.text for para in document.paragraphs])
            return content
        else:
            # Assume plain text for other file types
            with open(file_path, 'r', encoding='utf-8') as f:
                content = f.read()
            return content
    except FileNotFoundError:
        return f"Error: Document not found at {file_path}"
    except Exception as e:
        return f"Error reading document: {e}"

# Function to simulate 'docling' processing
def process_document_for_docling(document_content, analysis_recipe_name='Standard_GRC'):
    print(f"Processing document content for 'docling' using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for keyword extraction.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    keywords = recipe_to_use['keywords']
    found_keywords = [kw for kw in keywords if kw in document_content.lower()]

    processed_data = {
        'document_summary': document_content[:100],
        'extracted_keywords': found_keywords,
        'keyword_count': len(found_keywords),
        'recipe_used': analysis_recipe_name
    }
    print(f"Extracted keywords: {found_keywords}")
    return processed_data

# Function for GRC related operations
def perform_grc_checks(processed_data, analysis_recipe_name='Standard_GRC'):
    print(f"Performing GRC checks using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for GRC checks.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    grc_categories = recipe_to_use['grc_categories']
    extracted_keywords = processed_data.get('extracted_keywords', [])

    grc_findings = []
    categorized_findings = {}
    for category, category_keywords in grc_categories.items():
        for kw in category_keywords:
            if kw in extracted_keywords:
                if category not in categorized_findings:
                    categorized_findings[category] = []
                categorized_findings[category].append(f"Keyword '{kw}' detected in {category} area.")

    if categorized_findings:
        for category, findings in categorized_findings.items():
            grc_findings.append(f"{category} Findings: {'; '.join(findings)}")
    else:
        grc_findings.append(f"No specific GRC flags based on defined categories for {analysis_recipe_name}.")

    grc_report = {
        'summary': f"GRC analysis based on '{analysis_recipe_name}' recipe output.",
        'findings': grc_findings,
        'docling_output': processed_data,
        'recipe_used_for_checks': analysis_recipe_name
    }
    return grc_report

def main_grc_doc_processor(document_folder, analysis_recipe_name='Standard_GRC'):
    print(f"Starting GRC document processing in folder: {document_folder} with recipe: {analysis_recipe_name}")
    if not os.path.exists(document_folder):
        print(f"Folder not found: {document_folder}")
        return

    for filename in os.listdir(document_folder):
        file_path = os.path.join(document_folder, filename)
        if os.path.isfile(file_path):
            print(f"\n--- Processing {filename} with recipe {analysis_recipe_name} ---")
            doc_content = read_document(file_path)
            if not doc_content.startswith("Error"):
                docling_output = process_document_for_docling(doc_content, analysis_recipe_name)
                grc_output = perform_grc_checks(docling_output, analysis_recipe_name)
                print(f"Final GRC output for {filename}:\n{grc_output}")
            else:
                print(doc_content)

# Example usage (uncomment and modify to run):
document_directory = 'my_documents'
os.makedirs(document_directory, exist_ok=True)
# Create a dummy file for demonstration
with open(os.path.join(document_directory, 'sample_doc.txt'), 'w') as f:
    f.write("This is a sample document for docling and GRC processing. It contains some keywords like compliance and risk, privacy and GDPR.")

main_grc_doc_processor(document_directory, analysis_recipe_name='Standard_GRC')

print("\n--- Program structure ready ---")
print("Please replace the placeholder logic with your actual document processing and GRC requirements.")
print("Uncomment the example usage section and modify 'document_directory' to run it.")
```

```python
def calculate_grc_score(processed_data, analysis_recipe_name='Standard_GRC'):
    max_score = 100
    current_score = max_score
    default_deduction = 5 # Default weight for categories not explicitly listed

    print(f"Calculating GRC score for recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found for scoring. Using 'Standard_GRC'.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    grc_categories = recipe_to_use.get('grc_categories', {})
    category_weights = recipe_to_use.get('category_weights', {})
    extracted_keywords = processed_data.get('extracted_keywords', [])

    findings_categories = set()
    for category, category_keywords in grc_categories.items():
        for kw in category_keywords:
            if kw in extracted_keywords:
                findings_categories.add(category)
                break # Only need to find one keyword to mark category as having findings

    for category_with_finding in findings_categories:
        deduction = category_weights.get(category_with_finding, default_deduction)
        print(f"  - Found findings in category '{category_with_finding}', deducting {deduction} points.")
        current_score -= deduction
        if current_score < 0:
            current_score = 0
            break

    print(f"GRC Score: {current_score}/{max_score}")
    return current_score


import os
import PyPDF2 # For PDF files
import docx # For DOCX files

# Function to simulate document parsing/reading (from previous cells)
def read_document(file_path):
    # Re-using previous definition from cell 2da1d77c to avoid issues
    print(f"Reading document: {file_path}")
    try:
        if file_path.lower().endswith('.pdf'):
            with open(file_path, 'rb') as f:
                reader = PyPDF2.PdfReader(f)
                content = "".join([page.extract_text() for page in reader.pages if page.extract_text()])
            return content
        elif file_path.lower().endswith('.docx'):
            document = docx.Document(file_path)
            content = "\n".join([para.text for para in document.paragraphs])
            return content
        else:
            # Assume plain text for other file types
            with open(file_path, 'r', encoding='utf-8') as f:
                content = f.read()
            return content
    except FileNotFoundError:
        return f"Error: Document not found at {file_path}"
    except Exception as e:
        return f"Error reading document: {e}"

# Function to simulate 'docling' processing (from previous cells)
def process_document_for_docling(document_content, analysis_recipe_name='Standard_GRC'):
    # Re-using previous definition from cell 2da1d77c to avoid issues
    print(f"Processing document content for 'docling' using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for keyword extraction.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    keywords = recipe_to_use['keywords']
    found_keywords = [kw for kw in keywords if kw in document_content.lower()]

    processed_data = {
        'document_summary': document_content[:100],
        'extracted_keywords': found_keywords,
        'keyword_count': len(found_keywords),
        'recipe_used': analysis_recipe_name
    }
    print(f"Extracted keywords: {found_keywords}")
    return processed_data

# Function for GRC related operations (from previous cells)
def perform_grc_checks(processed_data, analysis_recipe_name='Standard_GRC'):
    # Re-using previous definition from cell 2da1d77c to avoid issues
    print(f"Performing GRC checks using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for GRC checks.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    grc_categories = recipe_to_use['grc_categories']
    extracted_keywords = processed_data.get('extracted_keywords', [])

    grc_findings = []
    categorized_findings = {}
    for category, category_keywords in grc_categories.items():
        for kw in category_keywords:
            if kw in extracted_keywords:
                if category not in categorized_findings:
                    categorized_findings[category] = []
                categorized_findings[category].append(f"Keyword '{kw}' detected in {category} area.")

    if categorized_findings:
        for category, findings in categorized_findings.items():
            grc_findings.append(f"{category} Findings: {'; '.join(findings)}")
    else:
        grc_findings.append(f"No specific GRC flags based on defined categories for {analysis_recipe_name}.")

    grc_report = {
        'summary': f"GRC analysis based on '{analysis_recipe_name}' recipe output.",
        'findings': grc_findings,
        'docling_output': processed_data,
        'recipe_used_for_checks': analysis_recipe_name
    }
    return grc_report

# Main GRC document processor function (updated to include score calculation)
def main_grc_doc_processor(document_folder, analysis_recipe_name='Standard_GRC'):
    print(f"Starting GRC document processing in folder: {document_folder} with recipe: {analysis_recipe_name}")
    if not os.path.exists(document_folder):
        print(f"Folder not found: {document_folder}")
        return

    for filename in os.listdir(document_folder):
        file_path = os.path.join(document_folder, filename)
        if os.path.isfile(file_path):
            print(f"\n--- Processing {filename} with recipe {analysis_recipe_name} ---")
            doc_content = read_document(file_path)
            if not doc_content.startswith("Error"):
                docling_output = process_document_for_docling(doc_content, analysis_recipe_name)
                grc_output = perform_grc_checks(docling_output, analysis_recipe_name)
                grc_score = calculate_grc_score(docling_output, analysis_recipe_name) # Calculate the score
                grc_output['grc_score'] = grc_score # Add score to the output
                print(f"Final GRC output for {filename}:\n{grc_output}")
            else:
                print(doc_content)

# Example usage to demonstrate scoring
document_directory = 'my_documents'
os.makedirs(document_directory, exist_ok=True)
with open(os.path.join(document_directory, 'sample_doc.txt'), 'w') as f:
    f.write("This document discusses compliance, risk management, data protection, privacy, and cybersecurity. It also mentions ethics.")

print("\n--- Demonstrating 'Standard_GRC' recipe with scoring ---")
main_grc_doc_processor(document_directory, analysis_recipe_name='Standard_GRC')

print("\n--- Demonstrating 'Privacy_Focused_GRC' recipe with scoring ---")
main_grc_doc_processor(document_directory, analysis_recipe_name='Privacy_Focused_GRC')

print("\n--- Demonstrating 'Security_Focused_GRC' recipe with scoring ---")
main_grc_doc_processor(document_directory, analysis_recipe_name='Security_Focused_GRC')
```

Create a new function, `format_grc_report`, that takes the raw `grc_output` dictionary (including the score and badge) and formats it into a human-readable string, separating presentation logic from analysis logic.

I will define the `format_grc_report` function as instructed, extracting the necessary data from the input dictionary and formatting it into a human-readable string.



```python
def format_grc_report(grc_output_dict):
    """
    Formats the raw GRC output dictionary into a human-readable string.

    Args:
        grc_output_dict (dict): The dictionary containing GRC analysis results,
                                 including score and badge.

    Returns:
        str: A multi-line string representing the formatted GRC report.
    """
    summary = grc_output_dict.get('summary', 'No summary available.')
    findings = grc_output_dict.get('findings', ['No findings to report.'])
    grc_score = grc_output_dict.get('grc_score', 'N/A')
    grc_badge = grc_output_dict.get('grc_badge', 'N/A')

    docling_output = grc_output_dict.get('docling_output', {})
    extracted_keywords = docling_output.get('extracted_keywords', [])
    recipe_used = docling_output.get('recipe_used', 'N/A')

    report_string = f"""
--- GRC Analysis Report ---
Summary: {summary}

GRC Score: {grc_score}/100
GRC Badge: {grc_badge}

Analysis Recipe Used: {recipe_used}
Extracted Keywords: {', '.join(extracted_keywords) if extracted_keywords else 'None'}

Findings:
"""

    for finding in findings:
        report_string += f"- {finding}\n"

    report_string += "---------------------------"

    return report_string

print("The `format_grc_report` function has been defined.")
```

**Reasoning**:
Now that the `format_grc_report` function is defined, I need to integrate it into the `main_grc_doc_processor` function. This involves updating `main_grc_doc_processor` to call `format_grc_report` with the `grc_output` dictionary and then printing the formatted string.



```python
import os
import PyPDF2 # For PDF files
import docx # For DOCX files

# Function to simulate document parsing/reading
def read_document(file_path):
    print(f"Reading document: {file_path}")
    try:
        if file_path.lower().endswith('.pdf'):
            with open(file_path, 'rb') as f:
                reader = PyPDF2.PdfReader(f)
                content = "".join([page.extract_text() for page in reader.pages if page.extract_text()])
            return content
        elif file_path.lower().endswith('.docx'):
            document = docx.Document(file_path)
            content = "\n".join([para.text for para in document.paragraphs])
            return content
        else:
            # Assume plain text for other file types
            with open(file_path, 'r', encoding='utf-8') as f:
                content = f.read()
            return content
    except FileNotFoundError:
        return f"Error: Document not found at {file_path}"
    except Exception as e:
        return f"Error reading document: {e}"

# Function to simulate 'docling' processing
def process_document_for_docling(document_content, analysis_recipe_name='Standard_GRC'):
    print(f"Processing document content for 'docling' using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for keyword extraction.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    keywords = recipe_to_use['keywords']
    found_keywords = [kw for kw in keywords if kw in document_content.lower()]

    processed_data = {
        'document_summary': document_content[:100],
        'extracted_keywords': found_keywords,
        'keyword_count': len(found_keywords),
        'recipe_used': analysis_recipe_name
    }
    print(f"Extracted keywords: {found_keywords}")
    return processed_data

# Function for GRC related operations
def perform_grc_checks(processed_data, analysis_recipe_name='Standard_GRC'):
    print(f"Performing GRC checks using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for GRC checks.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    grc_categories = recipe_to_use['grc_categories']
    extracted_keywords = processed_data.get('extracted_keywords', [])

    grc_findings = []
    categorized_findings = {}
    for category, category_keywords in grc_categories.items():
        for kw in category_keywords:
            if kw in extracted_keywords:
                if category not in categorized_findings:
                    categorized_findings[category] = []
                categorized_findings[category].append(f"Keyword '{kw}' detected in {category} area.")

    if categorized_findings:
        for category, findings in categorized_findings.items():
            grc_findings.append(f"{category} Findings: {'; '.join(findings)}")
    else:
        grc_findings.append(f"No specific GRC flags based on defined categories for {analysis_recipe_name}.")

    grc_report = {
        'summary': f"GRC analysis based on '{analysis_recipe_name}' recipe output.",
        'findings': grc_findings,
        'docling_output': processed_data,
        'recipe_used_for_checks': analysis_recipe_name
    }
    return grc_report

def calculate_grc_score(processed_data, analysis_recipe_name='Standard_GRC'):
    max_score = 100
    current_score = max_score
    default_deduction = 5 # Default weight for categories not explicitly listed

    print(f"Calculating GRC score for recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found for scoring. Using 'Standard_GRC'.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    grc_categories = recipe_to_use.get('grc_categories', {})
    category_weights = recipe_to_use.get('category_weights', {})
    extracted_keywords = processed_data.get('extracted_keywords', [])

    findings_categories = set()
    for category, category_keywords in grc_categories.items():
        for kw in category_keywords:
            if kw in extracted_keywords:
                findings_categories.add(category)
                break # Only need to find one keyword to mark category as having findings

    for category_with_finding in findings_categories:
        deduction = category_weights.get(category_with_finding, default_deduction)
        print(f"  - Found findings in category '{category_with_finding}', deducting {deduction} points.")
        current_score -= deduction
        if current_score < 0:
            current_score = 0
            break

    print(f"GRC Score: {current_score}/{max_score}")
    return current_score

def generate_grc_badge(grc_score):
    """
    Generates a textual representation of a GRC badge based on the GRC score.

    Args:
        grc_score (int): The calculated GRC score.

    Returns:
        str: The badge level (e.g., 'Gold', 'Silver', 'Bronze', 'None').
    """
    if grc_score >= 90:
        return 'Gold'
    elif grc_score >= 70:
        return 'Silver'
    elif grc_score >= 50:
        return 'Bronze'
    else:
        return 'None'

def format_grc_report(grc_output_dict):
    """
    Formats the raw GRC output dictionary into a human-readable string.

    Args:
        grc_output_dict (dict): The dictionary containing GRC analysis results,
                                 including score and badge.

    Returns:
        str: A multi-line string representing the formatted GRC report.
    """
    summary = grc_output_dict.get('summary', 'No summary available.')
    findings = grc_output_dict.get('findings', ['No findings to report.'])
    grc_score = grc_output_dict.get('grc_score', 'N/A')
    grc_badge = grc_output_dict.get('grc_badge', 'N/A')

    docling_output = grc_output_dict.get('docling_output', {})
    extracted_keywords = docling_output.get('extracted_keywords', [])
    recipe_used = docling_output.get('recipe_used', 'N/A')

    report_string = f"""
--- GRC Analysis Report ---
Summary: {summary}

GRC Score: {grc_score}/100
GRC Badge: {grc_badge}

Analysis Recipe Used: {recipe_used}
Extracted Keywords: {', '.join(extracted_keywords) if extracted_keywords else 'None'}

Findings:
"""

    for finding in findings:
        report_string += f"- {finding}\n"

    report_string += "---------------------------"

    return report_string

# Main GRC document processor function (updated to use format_grc_report)
def main_grc_doc_processor(document_folder, analysis_recipe_name='Standard_GRC'):
    print(f"Starting GRC document processing in folder: {document_folder} with recipe: {analysis_recipe_name}")
    if not os.path.exists(document_folder):
        print(f"Folder not found: {document_folder}")
        return

    for filename in os.listdir(document_folder):
        file_path = os.path.join(document_folder, filename)
        if os.path.isfile(file_path):
            print(f"\n--- Processing {filename} with recipe {analysis_recipe_name} ---")
            doc_content = read_document(file_path)
            if not doc_content.startswith("Error"):
                docling_output = process_document_for_docling(doc_content, analysis_recipe_name)
                grc_output = perform_grc_checks(docling_output, analysis_recipe_name)
                grc_score = calculate_grc_score(docling_output, analysis_recipe_name) # Calculate the score
                grc_badge = generate_grc_badge(grc_score) # Generate the badge
                grc_output['grc_score'] = grc_score # Add score to the output
                grc_output['grc_badge'] = grc_badge # Add badge to the output

                # Format the report using the new function
                formatted_report = format_grc_report(grc_output)
                print(f"Final GRC report for {filename}:\n{formatted_report}")
            else:
                print(doc_content)

# Example usage to demonstrate scoring and badge generation
document_directory = 'my_documents'
os.makedirs(document_directory, exist_ok=True)
with open(os.path.join(document_directory, 'sample_doc.txt'), 'w') as f:
    f.write("This document discusses compliance, risk management, data protection, privacy, and cybersecurity. It also mentions ethics.")

print("\n--- Demonstrating 'Standard_GRC' recipe with scoring and badge and formatted output ---")
main_grc_doc_processor(document_directory, analysis_recipe_name='Standard_GRC')

print("\n--- Demonstrating 'Privacy_Focused_GRC' recipe with scoring and badge and formatted output ---")
main_grc_doc_processor(document_directory, analysis_recipe_name='Privacy_Focused_GRC')

print("\n--- Demonstrating 'Security_Focused_GRC' recipe with scoring and badge and formatted output ---")
main_grc_doc_processor(document_directory, analysis_recipe_name='Security_Focused_GRC')
```

```python
import os
import PyPDF2 # For PDF files
import docx # For DOCX files

# Function to simulate document parsing/reading
def read_document(file_path):
    print(f"Reading document: {file_path}")
    try:
        if file_path.lower().endswith('.pdf'):
            with open(file_path, 'rb') as f:
                reader = PyPDF2.PdfReader(f)
                content = "".join([page.extract_text() for page in reader.pages if page.extract_text()])
            return content
        elif file_path.lower().endswith('.docx'):
            document = docx.Document(file_path)
            content = "\n".join([para.text for para in document.paragraphs])
            return content
        else:
            # Assume plain text for other file types
            with open(file_path, 'r', encoding='utf-8') as f:
                content = f.read()
            return content
    except FileNotFoundError:
        return f"Error: Document not found at {file_path}"
    except Exception as e:
        return f"Error reading document: {e}"

# Function to simulate 'docling' processing
def process_document_for_docling(document_content, analysis_recipe_name='Standard_GRC'):
    print(f"Processing document content for 'docling' using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for keyword extraction.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    keywords = recipe_to_use['keywords']
    found_keywords = [kw for kw in keywords if kw in document_content.lower()]

    processed_data = {
        'document_summary': document_content[:100],
        'extracted_keywords': found_keywords,
        'keyword_count': len(found_keywords),
        'recipe_used': analysis_recipe_name
    }
    print(f"Extracted keywords: {found_keywords}")
    return processed_data

# Function for GRC related operations
def perform_grc_checks(processed_data, analysis_recipe_name='Standard_GRC'):
    print(f"Performing GRC checks using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for GRC checks.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    grc_categories = recipe_to_use['grc_categories']
    extracted_keywords = processed_data.get('extracted_keywords', [])

    grc_findings = []
    categorized_findings = {}
    for category, category_keywords in grc_categories.items():
        for kw in category_keywords:
            if kw in extracted_keywords:
                if category not in categorized_findings:
                    categorized_findings[category] = []
                categorized_findings[category].append(f"Keyword '{kw}' detected in {category} area.")

    if categorized_findings:
        for category, findings in categorized_findings.items():
            grc_findings.append(f"{category} Findings: {'; '.join(findings)}")
    else:
        grc_findings.append(f"No specific GRC flags based on defined categories for {analysis_recipe_name}.")

    grc_report = {
        'summary': f"GRC analysis based on '{analysis_recipe_name}' recipe output.",
        'findings': grc_findings,
        'docling_output': processed_data,
        'recipe_used_for_checks': analysis_recipe_name
    }
    return grc_report

def calculate_grc_score(processed_data, analysis_recipe_name='Standard_GRC'):
    max_score = 100
    current_score = max_score
    default_deduction = 5 # Default weight for categories not explicitly listed

    print(f"Calculating GRC score for recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found for scoring. Using 'Standard_GRC'.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    grc_categories = recipe_to_use.get('grc_categories', {})
    category_weights = recipe_to_use.get('category_weights', {})
    extracted_keywords = processed_data.get('extracted_keywords', [])

    findings_categories = set()
    for category, category_keywords in grc_categories.items():
        for kw in category_keywords:
            if kw in extracted_keywords:
                findings_categories.add(category)
                break # Only need to find one keyword to mark category as having findings

    for category_with_finding in findings_categories:
        deduction = category_weights.get(category_with_finding, default_deduction)
        print(f"  - Found findings in category '{category_with_finding}', deducting {deduction} points.")
        current_score -= deduction
        if current_score < 0:
            current_score = 0
            break

    print(f"GRC Score: {current_score}/{max_score}")
    return current_score

def generate_grc_badge(grc_score):
    """
    Generates a textual representation of a GRC badge based on the GRC score.

    Args:
        grc_score (int): The calculated GRC score.

    Returns:
        str: The badge level (e.g., 'Gold', 'Silver', 'Bronze', 'None').
    """
    if grc_score >= 90:
        return 'Gold'
    elif grc_score >= 70:
        return 'Silver'
    elif grc_score >= 50:
        return 'Bronze'
    else:
        return 'None'

def format_grc_report(grc_output_dict,
                      include_summary=True,
                      include_score_badge=True,
                      include_keywords=True,
                      include_findings=True):
    """
    Formats the raw GRC output dictionary into a human-readable string,
    with configurable sections.

    Args:
        grc_output_dict (dict): The dictionary containing GRC analysis results,
                                 including score and badge.
        include_summary (bool): Whether to include the summary section.
        include_score_badge (bool): Whether to include the GRC score and badge section.
        include_keywords (bool): Whether to include the extracted keywords section.
        include_findings (bool): Whether to include the detailed findings section.

    Returns:
        str: A multi-line string representing the formatted GRC report.
    """
    report_string = """
--- GRC Analysis Report ---
"""

    if include_summary:
        summary = grc_output_dict.get('summary', 'No summary available.')
        report_string += f"Summary: {summary}\n\n"

    if include_score_badge:
        grc_score = grc_output_dict.get('grc_score', 'N/A')
        grc_badge = grc_output_dict.get('grc_badge', 'N/A')
        report_string += f"GRC Score: {grc_score}/100\n"
        report_string += f"GRC Badge: {grc_badge}\n\n"

    docling_output = grc_output_dict.get('docling_output', {})
    recipe_used = docling_output.get('recipe_used', 'N/A')
    report_string += f"Analysis Recipe Used: {recipe_used}\n"

    if include_keywords:
        extracted_keywords = docling_output.get('extracted_keywords', [])
        report_string += f"Extracted Keywords: {', '.join(extracted_keywords) if extracted_keywords else 'None'}\n\n"

    if include_findings:
        findings = grc_output_dict.get('findings', ['No findings to report.'])
        report_string += "Findings:\n"
        for finding in findings:
            report_string += f"- {finding}\n"

    report_string += "---------------------------"

    return report_string

# Main GRC document processor function (updated to use format_grc_report with customization options)
def main_grc_doc_processor(document_folder, analysis_recipe_name='Standard_GRC', output_file_path=None, **report_format_options):
    print(f"Starting GRC document processing in folder: {document_folder} with recipe: {analysis_recipe_name}")
    if not os.path.exists(document_folder):
        print(f"Folder not found: {document_folder}")
        return

    for filename in os.listdir(document_folder):
        file_path = os.path.join(document_folder, filename)
        if os.path.isfile(file_path):
            print(f"\n--- Processing {filename} with recipe {analysis_recipe_name} ---")
            doc_content = read_document(file_path)
            if not doc_content.startswith("Error"):
                docling_output = process_document_for_docling(doc_content, analysis_recipe_name)
                grc_output = perform_grc_checks(docling_output, analysis_recipe_name)
                grc_score = calculate_grc_score(docling_output, analysis_recipe_name) # Calculate the score
                grc_badge = generate_grc_badge(grc_score) # Generate the badge
                grc_output['grc_score'] = grc_score # Add score to the output
                grc_output['grc_badge'] = grc_badge # Add badge to the output

                # Format the report using the new function with custom options
                formatted_report = format_grc_report(grc_output, **report_format_options)

                if output_file_path:
                    output_dir = os.path.dirname(output_file_path)
                    if output_dir and not os.path.exists(output_dir):
                        os.makedirs(output_dir, exist_ok=True)
                    with open(output_file_path, 'w', encoding='utf-8') as f:
                        f.write(formatted_report)
                    print(f"Final GRC report for {filename} saved to {output_file_path}")
                else:
                    print(f"Final GRC report for {filename}:\n{formatted_report}")
            else:
                print(doc_content)

# Example usage to demonstrate scoring and badge generation
document_directory = 'my_documents'
os.makedirs(document_directory, exist_ok=True)
with open(os.path.join(document_directory, 'sample_doc.txt'), 'w') as f:
    f.write("This document discusses compliance, risk management, data protection, privacy, and cybersecurity. It also mentions ethics.")

print("\n--- Demonstrating 'Standard_GRC' recipe with full formatted output ---")
main_grc_doc_processor(document_directory, analysis_recipe_name='Standard_GRC')

print("\n--- Demonstrating 'Privacy_Focused_GRC' recipe with concise output (no keywords) ---")
main_grc_doc_processor(document_directory, analysis_recipe_name='Privacy_Focused_GRC', include_keywords=False)

print("\n--- Demonstrating 'Security_Focused_GRC' recipe with summary and score only ---")
main_grc_doc_processor(document_directory, analysis_recipe_name='Security_Focused_GRC', include_findings=False, include_keywords=False)

print("\n--- Demonstrating 'Standard_GRC' recipe with findings only ---")
main_grc_doc_processor(document_directory, analysis_recipe_name='Standard_GRC', include_summary=False, include_score_badge=False, include_keywords=False)

print("\n--- Demonstrating saving full report output to a file (Standard_GRC) ---")
output_report_file = os.path.join(document_directory, 'full_grc_report.txt')
main_grc_doc_processor(document_directory, analysis_recipe_name='Standard_GRC', output_file_path=output_report_file)
print(f"\nContent of {output_report_file}:")
with open(output_report_file, 'r', encoding='utf-8') as f:
    print(f.read())
```

```python
import os
import PyPDF2 # For PDF files
import docx # For DOCX files

# Function to simulate document parsing/reading
def read_document(file_path):
    print(f"Reading document: {file_path}")
    try:
        if file_path.lower().endswith('.pdf'):
            with open(file_path, 'rb') as f:
                reader = PyPDF2.PdfReader(f)
                content = "".join([page.extract_text() for page in reader.pages if page.extract_text()])
            return content
        elif file_path.lower().endswith('.docx'):
            document = docx.Document(file_path)
            content = "\n".join([para.text for para in document.paragraphs])
            return content
        else:
            # Assume plain text for other file types
            with open(file_path, 'r', encoding='utf-8') as f:
                content = f.read()
            return content
    except FileNotFoundError:
        return f"Error: Document not found at {file_path}"
    except Exception as e:
        return f"Error reading document: {e}"

# Function to simulate 'docling' processing
def process_document_for_docling(document_content, analysis_recipe_name='Standard_GRC'):
    print(f"Processing document content for 'docling' using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for keyword extraction.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    keywords = recipe_to_use['keywords']
    found_keywords = [kw for kw in keywords if kw in document_content.lower()]

    processed_data = {
        'document_summary': document_content[:100],
        'extracted_keywords': found_keywords,
        'keyword_count': len(found_keywords),
        'recipe_used': analysis_recipe_name
    }
    print(f"Extracted keywords: {found_keywords}")
    return processed_data

# Function for GRC related operations
def perform_grc_checks(processed_data, analysis_recipe_name='Standard_GRC'):
    print(f"Performing GRC checks using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for GRC checks.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    grc_categories = recipe_to_use['grc_categories']
    extracted_keywords = processed_data.get('extracted_keywords', [])

    grc_findings = []
    categorized_findings = {}
    for category, category_keywords in grc_categories.items():
        for kw in category_keywords:
            if kw in extracted_keywords:
                if category not in categorized_findings:
                    categorized_findings[category] = []
                categorized_findings[category].append(f"Keyword '{kw}' detected in {category} area.")

    if categorized_findings:
        for category, findings in categorized_findings.items():
            grc_findings.append(f"{category} Findings: {'; '.join(findings)}")
    else:
        grc_findings.append(f"No specific GRC flags based on defined categories for {analysis_recipe_name}.")

    grc_report = {
        'summary': f"GRC analysis based on '{analysis_recipe_name}' recipe output.",
        'findings': grc_findings,
        'docling_output': processed_data,
        'recipe_used_for_checks': analysis_recipe_name
    }
    return grc_report

def calculate_grc_score(processed_data, analysis_recipe_name='Standard_GRC'):
    max_score = 100
    current_score = max_score
    default_deduction = 5 # Default weight for categories not explicitly listed

    print(f"Calculating GRC score for recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found for scoring. Using 'Standard_GRC'.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    grc_categories = recipe_to_use.get('grc_categories', {})
    category_weights = recipe_to_use.get('category_weights', {})
    extracted_keywords = processed_data.get('extracted_keywords', [])

    findings_categories = set()
    for category, category_keywords in grc_categories.items():
        for kw in category_keywords:
            if kw in extracted_keywords:
                findings_categories.add(category)
                break # Only need to find one keyword to mark category as having findings

    for category_with_finding in findings_categories:
        deduction = category_weights.get(category_with_finding, default_deduction)
        print(f"  - Found findings in category '{category_with_finding}', deducting {deduction} points.")
        current_score -= deduction
        if current_score < 0:
            current_score = 0
            break

    print(f"GRC Score: {current_score}/{max_score}")
    return current_score

def generate_grc_badge(grc_score):
    """
    Generates a textual representation of a GRC badge based on the GRC score.

    Args:
        grc_score (int): The calculated GRC score.

    Returns:
        str: The badge level (e.g., 'Gold', 'Silver', 'Bronze', 'None').
    """
    if grc_score >= 90:
        return 'Gold'
    elif grc_score >= 70:
        return 'Silver'
    elif grc_score >= 50:
        return 'Bronze'
    else:
        return 'None'

def format_grc_report(grc_output_dict,
                      include_summary=True,
                      include_score_badge=True,
                      include_keywords=True,
                      include_findings=True,
                      output_format='text'):
    """
    Formats the raw GRC output dictionary into a human-readable string,
    with configurable sections and output format (text or html).

    Args:
        grc_output_dict (dict): The dictionary containing GRC analysis results,
                                 including score and badge.
        include_summary (bool): Whether to include the summary section.
        include_score_badge (bool): Whether to include the GRC score and badge section.
        include_keywords (bool): Whether to include the extracted keywords section.
        include_findings (bool): Whether to include the detailed findings section.
        output_format (str): The desired output format ('text' or 'html').

    Returns:
        str: A multi-line string representing the formatted GRC report in the specified format.
    """
    summary = grc_output_dict.get('summary', 'No summary available.')
    findings = grc_output_dict.get('findings', ['No findings to report.'])
    grc_score = grc_output_dict.get('grc_score', 'N/A')
    grc_badge = grc_output_dict.get('grc_badge', 'N/A')
    docling_output = grc_output_dict.get('docling_output', {})
    extracted_keywords = docling_output.get('extracted_keywords', [])
    recipe_used = docling_output.get('recipe_used', 'N/A')

    if output_format == 'html':
        report_string = """<!DOCTYPE html>
<html>
<head>
<title>GRC Analysis Report</title>
<style>
body { font-family: Arial, sans-serif; line-height: 1.6; color: #333; }
h1 { color: #2c3e50; border-bottom: 2px solid #3498db; padding-bottom: 10px; }
h2 { color: #34495e; }
ul { list-style-type: disc; margin-left: 20px; }
li { margin-bottom: 5px; }
.score-badge { font-weight: bold; color: #e74c3c; }
.gold { color: #ffd700; }
.silver { color: #c0c0c0; }
.bronze { color: #cd7f32; }
</style>
</head>
<body>
<h1>GRC Analysis Report</h1>
"""

        if include_summary:
            report_string += f"<h2>Summary</h2><p>{summary}</p>\n"

        if include_score_badge:
            badge_class = grc_badge.lower() if grc_badge != 'N/A' and grc_badge != 'None' else ''
            report_string += f"<h2>GRC Score & Badge</h2><p>GRC Score: <strong>{grc_score}/100</strong><br>GRC Badge: <span class=\"score-badge {badge_class}\">{grc_badge}</span></p>\n"

        report_string += f"<h2>Analysis Details</h2><p>Analysis Recipe Used: <strong>{recipe_used}</strong></p>\n"

        if include_keywords:
            report_string += f"<h3>Extracted Keywords:</h3>\n<ul>\n"
            if extracted_keywords:
                for kw in extracted_keywords:
                    report_string += f"<li>{kw}</li>\n"
            else:
                report_string += "<li>None</li>\n"
            report_string += "</ul>\n"

        if include_findings:
            report_string += f"<h3>Findings:</h3>\n<ul>\n"
            if findings:
                for finding in findings:
                    report_string += f"<li>{finding}</li>\n"
            else:
                report_string += "<li>No findings to report.</li>\n"
            report_string += "</ul>\n"

        report_string += "</body>\n</html>"

    else:  # Default to 'text' format
        report_string = f"""
--- GRC Analysis Report ---
"""

        if include_summary:
            report_string += f"Summary: {summary}\n\n"

        if include_score_badge:
            report_string += f"GRC Score: {grc_score}/100\n"
            report_string += f"GRC Badge: {grc_badge}\n\n"

        report_string += f"Analysis Recipe Used: {recipe_used}\n"

        if include_keywords:
            report_string += f"Extracted Keywords: {', '.join(extracted_keywords) if extracted_keywords else 'None'}\n\n"

        if include_findings:
            report_string += "Findings:\n"
            for finding in findings:
                report_string += f"- {finding}\n"

        report_string += "---------------------------"

    return report_string

# Main GRC document processor function (updated to use format_grc_report with customization options)
def main_grc_doc_processor(document_folder, analysis_recipe_name='Standard_GRC', output_file_path=None, output_format='text', **report_format_options):
    print(f"Starting GRC document processing in folder: {document_folder} with recipe: {analysis_recipe_name}")
    if not os.path.exists(document_folder):
        print(f"Folder not found: {document_folder}")
        return

    for filename in os.listdir(document_folder):
        file_path = os.path.join(document_folder, filename)
        if os.path.isfile(file_path):
            print(f"\n--- Processing {filename} with recipe {analysis_recipe_name} ---")
            doc_content = read_document(file_path)
            if not doc_content.startswith("Error"):
                docling_output = process_document_for_docling(doc_content, analysis_recipe_name)
                grc_output = perform_grc_checks(docling_output, analysis_recipe_name)
                grc_score = calculate_grc_score(docling_output, analysis_recipe_name) # Calculate the score
                grc_badge = generate_grc_badge(grc_score) # Generate the badge
                grc_output['grc_score'] = grc_score # Add score to the output
                grc_output['grc_badge'] = grc_badge # Add badge to the output

                # Format the report using the new function with custom options and output format
                formatted_report = format_grc_report(grc_output, output_format=output_format, **report_format_options)

                if output_file_path:
                    base_name, _ = os.path.splitext(output_file_path)
                    if output_format == 'html':
                        final_output_path = f"{base_name}.html"
                    else:
                        final_output_path = f"{base_name}.txt"

                    output_dir = os.path.dirname(final_output_path)
                    if output_dir and not os.path.exists(output_dir):
                        os.makedirs(output_dir, exist_ok=True)
                    with open(final_output_path, 'w', encoding='utf-8') as f:
                        f.write(formatted_report)
                    print(f"Final GRC report for {filename} saved to {final_output_path}")
                else:
                    print(f"Final GRC report for {filename}:\n{formatted_report}")
            else:
                print(doc_content)

# Example usage to demonstrate scoring and badge generation
document_directory = 'my_documents'
os.makedirs(document_directory, exist_ok=True)
with open(os.path.join(document_directory, 'sample_doc.txt'), 'w') as f:
    f.write("This document discusses compliance, risk management, data protection, privacy, and cybersecurity. It also mentions ethics.")

print("\n--- Demonstrating 'Standard_GRC' recipe with full formatted text output ---")
main_grc_doc_processor(document_directory, analysis_recipe_name='Standard_GRC', output_format='text')

print("\n--- Demonstrating 'Privacy_Focused_GRC' recipe with concise text output (no keywords) ---")
main_grc_doc_processor(document_directory, analysis_recipe_name='Privacy_Focused_GRC', output_format='text', include_keywords=False)

print("\n--- Demonstrating 'Security_Focused_GRC' recipe with summary and score only (text output) ---")
main_grc_doc_processor(document_directory, analysis_recipe_name='Security_Focused_GRC', output_format='text', include_findings=False, include_keywords=False)

print("\n--- Demonstrating saving full report output to a text file (Standard_GRC) ---")
output_report_text_file = os.path.join(document_directory, 'full_grc_report.txt')
main_grc_doc_processor(document_directory, analysis_recipe_name='Standard_GRC', output_file_path=output_report_text_file, output_format='text')
print(f"\nContent of {output_report_text_file}:")
with open(output_report_text_file, 'r', encoding='utf-8') as f:
    print(f.read())

print("\n--- Demonstrating saving full report output to an HTML file (Privacy_Focused_GRC) ---")
output_report_html_file = os.path.join(document_directory, 'privacy_grc_report.html')
main_grc_doc_processor(document_directory, analysis_recipe_name='Privacy_Focused_GRC', output_file_path=output_report_html_file, output_format='html')
print(f"\nContent of {output_report_html_file}: (showing first 500 chars)\n")
with open(output_report_html_file, 'r', encoding='utf-8') as f:
    print(f.read()[:500])
```

```python
import os
import PyPDF2 # For PDF files
import docx # For DOCX files
import matplotlib.pyplot as plt

# Function to simulate document parsing/reading
def read_document(file_path):
    print(f"Reading document: {file_path}")
    try:
        if file_path.lower().endswith('.pdf'):
            with open(file_path, 'rb') as f:
                reader = PyPDF2.PdfReader(f)
                content = "".join([page.extract_text() for page in reader.pages if page.extract_text()])
            return content
        elif file_path.lower().endswith('.docx'):
            document = docx.Document(file_path)
            content = "\n".join([para.text for para in document.paragraphs])
            return content
        else:
            # Assume plain text for other file types
            with open(file_path, 'r', encoding='utf-8') as f:
                content = f.read()
            return content
    except FileNotFoundError:
        return f"Error: Document not found at {file_path}"
    except Exception as e:
        return f"Error reading document: {e}"

# Function to simulate 'docling' processing
def process_document_for_docling(document_content, analysis_recipe_name='Standard_GRC'):
    print(f"Processing document content for 'docling' using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for keyword extraction.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    keywords = recipe_to_use['keywords']
    found_keywords = [kw for kw in keywords if kw in document_content.lower()]

    processed_data = {
        'document_summary': document_content[:100],
        'extracted_keywords': found_keywords,
        'keyword_count': len(found_keywords),
        'recipe_used': analysis_recipe_name
    }
    print(f"Extracted keywords: {found_keywords}")
    return processed_data

# Function for GRC related operations
def perform_grc_checks(processed_data, analysis_recipe_name='Standard_GRC'):
    print(f"Performing GRC checks using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for GRC checks.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    grc_categories = recipe_to_use['grc_categories']
    extracted_keywords = processed_data.get('extracted_keywords', [])

    grc_findings = []
    categorized_findings = {}
    for category, category_keywords in grc_categories.items():
        for kw in category_keywords:
            if kw in extracted_keywords:
                if category not in categorized_findings:
                    categorized_findings[category] = []
                categorized_findings[category].append(f"Keyword '{kw}' detected in {category} area.")

    if categorized_findings:
        for category, findings in categorized_findings.items():
            grc_findings.append(f"{category} Findings: {'; '.join(findings)}")
    else:
        grc_findings.append(f"No specific GRC flags based on defined categories for {analysis_recipe_name}.")

    grc_report = {
        'summary': f"GRC analysis based on '{analysis_recipe_name}' recipe output.",
        'findings': grc_findings,
        'docling_output': processed_data,
        'recipe_used_for_checks': analysis_recipe_name
    }
    return grc_report

def calculate_grc_score(processed_data, analysis_recipe_name='Standard_GRC'):
    max_score = 100
    current_score = max_score
    default_deduction = 5 # Default weight for categories not explicitly listed

    print(f"Calculating GRC score for recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found for scoring. Using 'Standard_GRC'.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    grc_categories = recipe_to_use.get('grc_categories', {})
    category_weights = recipe_to_use.get('category_weights', {})
    extracted_keywords = processed_data.get('extracted_keywords', [])

    findings_categories = set()
    for category, category_keywords in grc_categories.items():
        for kw in category_keywords:
            if kw in extracted_keywords:
                findings_categories.add(category)
                break # Only need to find one keyword to mark category as having findings

    for category_with_finding in findings_categories:
        deduction = category_weights.get(category_with_finding, default_deduction)
        print(f"  - Found findings in category '{category_with_finding}', deducting {deduction} points.")
        current_score -= deduction
        if current_score < 0:
            current_score = 0
            break

    print(f"GRC Score: {current_score}/{max_score}")
    return current_score

def generate_grc_badge(grc_score):
    """
    Generates a textual representation of a GRC badge based on the GRC score.

    Args:
        grc_score (int): The calculated GRC score.

    Returns:
        str: The badge level (e.g., 'Gold', 'Silver', 'Bronze', 'None').
    """
    if grc_score >= 90:
        return 'Gold'
    elif grc_score >= 70:
        return 'Silver'
    elif grc_score >= 50:
        return 'Bronze'
    else:
        return 'None'

def generate_grc_visualizations(grc_output_dict, output_dir):
    """
    Generates visual representations (bar chart, pie chart) from GRC output data.

    Args:
        grc_output_dict (dict): The dictionary containing GRC analysis results.
        output_dir (str): The directory to save the generated plots.

    Returns:
        list: A list of file paths to the saved visualization images.
    """
    os.makedirs(output_dir, exist_ok=True)
    visualization_paths = []

    # Extract data for visualizations
    findings = grc_output_dict.get('findings', [])
    extracted_keywords = grc_output_dict.get('docling_output', {}).get('extracted_keywords', [])
    analysis_recipe_name = grc_output_dict.get('docling_output', {}).get('recipe_used', 'N/A')

    # --- Prepare data for Bar Chart (Findings per Category) ---
    findings_counts = {}
    for finding_str in findings:
        if ':' in finding_str:
            category_name = finding_str.split(':')[0].strip()
            if ' Findings' in category_name:
                category_name = category_name.replace(' Findings', '')
            findings_counts[category_name] = findings_counts.get(category_name, 0) + 1

    if findings_counts:
        categories = list(findings_counts.keys())
        counts = list(findings_counts.values())

        plt.figure(figsize=(10, 6))
        plt.bar(categories, counts, color='skyblue')
        plt.xlabel('GRC Category')
        plt.ylabel('Number of Findings')
        plt.title(f'Number of GRC Findings per Category ({analysis_recipe_name})')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        bar_chart_path = os.path.join(output_dir, f'{analysis_recipe_name.lower().replace(" ", "_")}_grc_findings_bar_chart.png')
        plt.savefig(bar_chart_path)
        plt.close()
        visualization_paths.paths.append(bar_chart_path)

    # --- Prepare data for Pie Chart (Keyword Distribution) ---
    if extracted_keywords:
        keyword_counts = {kw: extracted_keywords.count(kw) for kw in set(extracted_keywords)}

        labels = list(keyword_counts.keys())
        sizes = list(keyword_counts.values())

        plt.figure(figsize=(10, 7))
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, textprops={'fontsize': 9})
        plt.axis('equal')
        plt.title(f'Distribution of Extracted Keywords ({analysis_recipe_name})')
        plt.tight_layout()
        pie_chart_path = os.path.join(output_dir, f'{analysis_recipe_name.lower().replace(" ", "_")}_grc_keywords_pie_chart.png')
        plt.savefig(pie_chart_path)
        plt.close()
        visualization_paths.paths.append(pie_chart_path)

    print(f"Generated visualizations: {visualization_paths}")
    return visualization_paths

def format_grc_report(grc_output_dict,
                      include_summary=True,
                      include_score_badge=True,
                      include_keywords=True,
                      include_findings=True,
                      include_visualizations=True,
                      output_format='text'):
    """
    Formats the raw GRC output dictionary into a human-readable string,
    with configurable sections and output format (text or html).

    Args:
        grc_output_dict (dict): The dictionary containing GRC analysis results,
                                 including score and badge.
        include_summary (bool): Whether to include the summary section.
        include_score_badge (bool): Whether to include the GRC score and badge section.
        include_keywords (bool): Whether to include the extracted keywords section.
        include_findings (bool): Whether to include the detailed findings section.
        include_visualizations (bool): Whether to include links/references to visualizations.
        output_format (str): The desired output format ('text' or 'html').

    Returns:
        str: A multi-line string representing the formatted GRC report in the specified format.
    """
    summary = grc_output_dict.get('summary', 'No summary available.')
    findings = grc_output_dict.get('findings', ['No findings to report.'])
    grc_score = grc_output_dict.get('grc_score', 'N/A')
    grc_badge = grc_output_dict.get('grc_badge', 'N/A')
    docling_output = grc_output_dict.get('docling_output', {})
    extracted_keywords = docling_output.get('extracted_keywords', [])
    recipe_used = docling_output.get('recipe_used', 'N/A')
    visualization_paths = grc_output_dict.get('visualization_paths', [])

    if output_format == 'html':
        report_string = """<!DOCTYPE html>
<html>
<head>
<title>GRC Analysis Report</title>
<style>
body { font-family: Arial, sans-serif; line-height: 1.6; color: #333; }
h1 { color: #2c3e50; border-bottom: 2px solid #3498db; padding-bottom: 10px; }
h2 { color: #34495e; }
ul { list-style-type: disc; margin-left: 20px; }
li { margin-bottom: 5px; }
.score-badge { font-weight: bold; color: #e74c3c; }
.gold { color: #ffd700; }
.silver { color: #c0c0c0; }
.bronze { color: #cd7f32; }
img { max-width: 100%; height: auto; margin-top: 10px; border: 1px solid #ddd; }
</style>
</head>
<body>
<h1>GRC Analysis Report</h1>
"""

        if include_summary:
            report_string += f"<h2>Summary</h2><p>{summary}</p>\n"

        if include_score_badge:
            badge_class = grc_badge.lower() if grc_badge != 'N/A' and grc_badge != 'None' else ''
            report_string += f"<h2>GRC Score & Badge</h2><p>GRC Score: <strong>{grc_score}/100</strong><br>GRC Badge: <span class=\"score-badge {badge_class}\">{grc_badge}</span></p>\n"

        report_string += f"<h2>Analysis Details</h2><p>Analysis Recipe Used: <strong>{recipe_used}</strong></p>\n"

        if include_keywords:
            report_string += f"<h3>Extracted Keywords:</h3>\n<ul>\n"
            if extracted_keywords:
                for kw in extracted_keywords:
                    report_string += f"<li>{kw}</li>\n"
            else:
                report_string += "<li>None</li>\n"
            report_string += "</ul>\n"

        if include_findings:
            report_string += f"<h3>Findings:</h3>\n<ul>\n"
            if findings:
                for finding in findings:
                    report_string += f"<li>{finding}</li>\n"
            else:
                report_string += "<li>No findings to report.</li>\n"
            report_string += "</ul>\n"

        if include_visualizations and visualization_paths:
            report_string += "<h3>Visualizations:</h3>\n"
            for viz_path in visualization_paths:
                # Assuming viz_path is relative to the HTML report for local viewing
                # For Colab, paths like 'my_documents/std_grc_bar_chart.png' will work if HTML is in my_documents
                report_string += f"<img src='{os.path.basename(viz_path)}' alt='GRC Visualization'><br>\n"

        report_string += "</body>\n</html>"

    else:  # Default to 'text' format
        report_string = f"""
--- GRC Analysis Report ---
"""

        if include_summary:
            report_string += f"Summary: {summary}\n\n"

        if include_score_badge:
            report_string += f"GRC Score: {grc_score}/100\n"
            report_string += f"GRC Badge: {grc_badge}\n\n"

        report_string += f"Analysis Recipe Used: {recipe_used}\n"

        if include_keywords:
            report_string += f"Extracted Keywords: {', '.join(extracted_keywords) if extracted_keywords else 'None'}\n\n"

        if include_findings:
            report_string += "Findings:\n"
            for finding in findings:
                report_string += f"- {finding}\n"

        if include_visualizations and visualization_paths:
            report_string += "\nVisualizations:\n"
            for viz_path in visualization_paths:
                report_string += f"- {viz_path}\n"

        report_string += "---------------------------"

    return report_string

# Main GRC document processor function (updated to use format_grc_report with customization options)
def main_grc_doc_processor(document_folder, analysis_recipe_name='Standard_GRC', output_file_path=None, output_format='text', output_visualization_dir='grc_visualizations', **report_format_options):
    print(f"Starting GRC document processing in folder: {document_folder} with recipe: {analysis_recipe_name}")
    if not os.path.exists(document_folder):
        print(f"Folder not found: {document_folder}")
        return

    for filename in os.listdir(document_folder):
        file_path = os.path.join(document_folder, filename)
        if os.path.isfile(file_path):
            print(f"\n--- Processing {filename} with recipe {analysis_recipe_name} ---")
            doc_content = read_document(file_path)
            if not doc_content.startswith("Error"):
                docling_output = process_document_for_docling(doc_content, analysis_recipe_name)
                grc_output = perform_grc_checks(docling_output, analysis_recipe_name)
                grc_score = calculate_grc_score(docling_output, analysis_recipe_name) # Calculate the score
                grc_badge = generate_grc_badge(grc_score) # Generate the badge
                grc_output['grc_score'] = grc_score # Add score to the output
                grc_output['grc_badge'] = grc_badge # Add badge to the output

                # Generate visualizations and add their paths to grc_output
                full_output_viz_dir = os.path.join(document_folder, output_visualization_dir)
                visualization_paths = generate_grc_visualizations(grc_output, full_output_viz_dir)
                grc_output['visualization_paths'] = visualization_paths

                # Format the report using the new function with custom options and output format
                formatted_report = format_grc_report(grc_output, output_format=output_format, **report_format_options)

                if output_file_path:
                    base_name, _ = os.path.splitext(output_file_path)
                    if output_format == 'html':
                        final_output_path = f"{base_name}.html"
                    else:
                        final_output_path = f"{base_name}.txt"

                    output_dir = os.path.dirname(final_output_path)
                    if output_dir and not os.path.exists(output_dir):
                        os.makedirs(output_dir, exist_ok=True)
                    with open(final_output_path, 'w', encoding='utf-8') as f:
                        f.write(formatted_report)
                    print(f"Final GRC report for {filename} saved to {final_output_path}")
                else:
                    print(f"Final GRC report for {filename}:\n{formatted_report}")
            else:
                print(doc_content)

# Example usage to demonstrate scoring and badge generation
document_directory = 'my_documents'
os.makedirs(document_directory, exist_ok=True)
# Create a dummy file for demonstration, now including 'ethics'
with open(os.path.join(document_directory, 'sample_doc.txt'), 'w') as f:
    f.write("This document discusses compliance, risk management, data protection, privacy, and cybersecurity. It also mentions ethics.")

# --- Demonstrating 'Standard_GRC' recipe with full formatted HTML output including visualizations ---
print("\n--- Demonstrating 'Standard_GRC' recipe with full formatted HTML output including visualizations ---")
output_html_file_with_viz = os.path.join(document_directory, 'standard_grc_report_with_viz.html')
main_grc_doc_processor(document_directory, analysis_recipe_name='Standard_GRC', output_file_path=output_html_file_with_viz, output_format='html', include_visualizations=True)

print(f"\nContent of {output_html_file_with_viz}: (showing first 1000 chars)\n")
if os.path.exists(output_html_file_with_viz):
    with open(output_html_file_with_viz, 'r', encoding='utf-8') as f:
        print(f.read()[:1000])
else:
    print(f"File not found: {output_html_file_with_viz}")

# --- Demonstrating 'Privacy_Focused_GRC' recipe with full formatted HTML output including visualizations ---
print("\n--- Demonstrating 'Privacy_Focused_GRC' recipe with full formatted HTML output including visualizations ---")
output_html_file_with_viz_privacy = os.path.join(document_directory, 'privacy_grc_report_with_viz.html')
main_grc_doc_processor(document_directory, analysis_recipe_name='Privacy_Focused_GRC', output_file_path=output_html_file_with_viz_privacy, output_format='html', include_visualizations=True)

print(f"\nContent of {output_html_file_with_viz_privacy}: (showing first 1000 chars)\n")
if os.path.exists(output_html_file_with_viz_privacy):
    with open(output_html_file_with_viz_privacy, 'r', encoding='utf-8') as f:
        print(f.read()[:1000])
else:
    print(f"File not found: {output_html_file_with_viz_privacy}")
```

```python
import os
import PyPDF2 # For PDF files
import docx # For DOCX files
import matplotlib.pyplot as plt

# Function to simulate document parsing/reading
def read_document(file_path):
    print(f"Reading document: {file_path}")
    try:
        if file_path.lower().endswith('.pdf'):
            with open(file_path, 'rb') as f:
                reader = PyPDF2.PdfReader(f)
                content = "".join([page.extract_text() for page in reader.pages if page.extract_text()])
            return content
        elif file_path.lower().endswith('.docx'):
            document = docx.Document(file_path)
            content = "\n".join([para.text for para in document.paragraphs])
            return content
        else:
            # Assume plain text for other file types
            with open(file_path, 'r', encoding='utf-8') as f:
                content = f.read()
            return content
    except FileNotFoundError:
        return f"Error: Document not found at {file_path}"
    except Exception as e:
        return f"Error reading document: {e}"

# Function to simulate 'docling' processing
def process_document_for_docling(document_content, analysis_recipe_name='Standard_GRC'):
    print(f"Processing document content for 'docling' using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for keyword extraction.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    keywords = recipe_to_use['keywords']
    found_keywords = [kw for kw in keywords if kw in document_content.lower()]

    processed_data = {
        'document_summary': document_content[:100],
        'extracted_keywords': found_keywords,
        'keyword_count': len(found_keywords),
        'recipe_used': analysis_recipe_name
    }
    print(f"Extracted keywords: {found_keywords}")
    return processed_data

# Function for GRC related operations
def perform_grc_checks(processed_data, analysis_recipe_name='Standard_GRC'):
    print(f"Performing GRC checks using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for GRC checks.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    grc_categories = recipe_to_use['grc_categories']
    extracted_keywords = processed_data.get('extracted_keywords', [])

    grc_findings = []
    categorized_findings = {}
    for category, category_keywords in grc_categories.items():
        for kw in category_keywords:
            if kw in extracted_keywords:
                if category not in categorized_findings:
                    categorized_findings[category] = []
                categorized_findings[category].append(f"Keyword '{kw}' detected in {category} area.")

    if categorized_findings:
        for category, findings in categorized_findings.items():
            grc_findings.append(f"{category} Findings: {'; '.join(findings)}")
    else:
        grc_findings.append(f"No specific GRC flags based on defined categories for {analysis_recipe_name}.")

    grc_report = {
        'summary': f"GRC analysis based on '{analysis_recipe_name}' recipe output.",
        'findings': grc_findings,
        'docling_output': processed_data,
        'recipe_used_for_checks': analysis_recipe_name
    }
    return grc_report

def calculate_grc_score(processed_data, analysis_recipe_name='Standard_GRC'):
    max_score = 100
    current_score = max_score
    default_deduction = 5 # Default weight for categories not explicitly listed

    print(f"Calculating GRC score for recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found for scoring. Using 'Standard_GRC'.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    grc_categories = recipe_to_use.get('grc_categories', {})
    category_weights = recipe_to_use.get('category_weights', {})
    extracted_keywords = processed_data.get('extracted_keywords', [])

    findings_categories = set()
    for category, category_keywords in grc_categories.items():
        for kw in category_keywords:
            if kw in extracted_keywords:
                findings_categories.add(category)
                break # Only need to find one keyword to mark category as having findings

    for category_with_finding in findings_categories:
        deduction = category_weights.get(category_with_finding, default_deduction)
        print(f"  - Found findings in category '{category_with_finding}', deducting {deduction} points.")
        current_score -= deduction
        if current_score < 0:
            current_score = 0
            break

    print(f"GRC Score: {current_score}/{max_score}")
    return current_score

def generate_grc_badge(grc_score):
    """
    Generates a textual representation of a GRC badge based on the GRC score.

    Args:
        grc_score (int): The calculated GRC score.

    Returns:
        str: The badge level (e.g., 'Gold', 'Silver', 'Bronze', 'None').
    """
    if grc_score >= 90:
        return 'Gold'
    elif grc_score >= 70:
        return 'Silver'
    elif grc_score >= 50:
        return 'Bronze'
    else:
        return 'None'

def generate_grc_visualizations(grc_output_dict, output_dir):
    """
    Generates visual representations (bar chart, pie chart) from GRC output data.

    Args:
        grc_output_dict (dict): The dictionary containing GRC analysis results.
        output_dir (str): The directory to save the generated plots.

    Returns:
        list: A list of file paths to the saved visualization images.
    """
    os.makedirs(output_dir, exist_ok=True)
    visualization_paths = []

    # Extract data for visualizations
    findings = grc_output_dict.get('findings', [])
    extracted_keywords = grc_output_dict.get('docling_output', {}).get('extracted_keywords', [])
    analysis_recipe_name = grc_output_dict.get('docling_output', {}).get('recipe_used', 'N/A')

    # --- Prepare data for Bar Chart (Findings per Category) ---
    findings_counts = {}
    for finding_str in findings:
        if ':' in finding_str:
            category_name = finding_str.split(':')[0].strip()
            if ' Findings' in category_name:
                category_name = category_name.replace(' Findings', '')
            findings_counts[category_name] = findings_counts.get(category_name, 0) + 1

    if findings_counts:
        categories = list(findings_counts.keys())
        counts = list(findings_counts.values())

        plt.figure(figsize=(10, 6))
        plt.bar(categories, counts, color='skyblue')
        plt.xlabel('GRC Category')
        plt.ylabel('Number of Findings')
        plt.title(f'Number of GRC Findings per Category ({analysis_recipe_name})')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        bar_chart_path = os.path.join(output_dir, f'{analysis_recipe_name.lower().replace(" ", "_")}_grc_findings_bar_chart.png')
        plt.savefig(bar_chart_path)
        plt.close()
        visualization_paths.append(bar_chart_path)

    # --- Prepare data for Pie Chart (Keyword Distribution) ---
    if extracted_keywords:
        keyword_counts = {kw: extracted_keywords.count(kw) for kw in set(extracted_keywords)}

        labels = list(keyword_counts.keys())
        sizes = list(keyword_counts.values())

        plt.figure(figsize=(10, 7))
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, textprops={'fontsize': 9})
        plt.axis('equal')
        plt.title(f'Distribution of Extracted Keywords ({analysis_recipe_name})')
        plt.tight_layout()
        pie_chart_path = os.path.join(output_dir, f'{analysis_recipe_name.lower().replace(" ", "_")}_grc_keywords_pie_chart.png')
        plt.savefig(pie_chart_path)
        plt.close()
        visualization_paths.append(pie_chart_path)

    print(f"Generated visualizations: {visualization_paths}")
    return visualization_paths

def format_grc_report(grc_output_dict,
                      include_summary=True,
                      include_score_badge=True,
                      include_keywords=True,
                      include_findings=True,
                      include_visualizations=True,
                      output_format='markdown'): # Changed default to markdown
    """
    Formats the raw GRC output dictionary into a human-readable string,
    with configurable sections and output format (text, html, or markdown).

    Args:
        grc_output_dict (dict): The dictionary containing GRC analysis results,
                                 including score and badge.
        include_summary (bool): Whether to include the summary section.
        include_score_badge (bool): Whether to include the GRC score and badge section.
        include_keywords (bool): Whether to include the extracted keywords section.
        include_findings (bool): Whether to include the detailed findings section.
        include_visualizations (bool): Whether to include links/references to visualizations.
        output_format (str): The desired output format ('text', 'html', or 'markdown').

    Returns:
        str: A multi-line string representing the formatted GRC report in the specified format.
    """
    summary = grc_output_dict.get('summary', 'No summary available.')
    findings = grc_output_dict.get('findings', ['No findings to report.'])
    grc_score = grc_output_dict.get('grc_score', 'N/A')
    grc_badge = grc_output_dict.get('grc_badge', 'N/A')
    docling_output = grc_output_dict.get('docling_output', {})
    extracted_keywords = docling_output.get('extracted_keywords', [])
    recipe_used = docling_output.get('recipe_used', 'N/A')
    visualization_paths = grc_output_dict.get('visualization_paths', [])

    if output_format == 'html':
        report_string = """<!DOCTYPE html>
<html>
<head>
<title>GRC Analysis Report</title>
<style>
body { font-family: Arial, sans-serif; line-height: 1.6; color: #333; }
h1 { color: #2c3e50; border-bottom: 2px solid #3498db; padding-bottom: 10px; }
h2 { color: #34495e; }
ul { list-style-type: disc; margin-left: 20px; }
li { margin-bottom: 5px; }
.score-badge { font-weight: bold; color: #e74c3c; }
.gold { color: #ffd700; }
.silver { color: #c0c0c0; }
.bronze { color: #cd7f32; }
img { max-width: 100%; height: auto; margin-top: 10px; border: 1px solid #ddd; }
</style>
</head>
<body>
<h1>GRC Analysis Report</h1>
"""

        if include_summary:
            report_string += f"<h2>Summary</h2><p>{summary}</p>\n"

        if include_score_badge:
            badge_class = grc_badge.lower() if grc_badge != 'N/A' and grc_badge != 'None' else ''
            report_string += f"<h2>GRC Score & Badge</h2><p>GRC Score: <strong>{grc_score}/100</strong><br>GRC Badge: <span class=\"score-badge {badge_class}\">{grc_badge}</span></p>\n"

        report_string += f"<h2>Analysis Details</h2><p>Analysis Recipe Used: <strong>{recipe_used}</strong></p>\n"

        if include_keywords:
            report_string += f"<h3>Extracted Keywords:</h3>\n<ul>\n"
            if extracted_keywords:
                for kw in extracted_keywords:
                    report_string += f"<li>{kw}</li>\n"
            else:
                report_string += "<li>None</li>\n"
            report_string += "</ul>\n"

        if include_findings:
            report_string += f"<h3>Findings:</h3>\n<ul>\n"
            if findings:
                for finding in findings:
                    report_string += f"<li>{finding}</li>\n"
            else:
                report_string += "<li>No findings to report.</li>\n"
            report_string += "</ul>\n"

        if include_visualizations and visualization_paths:
            report_string += "<h3>Visualizations:</h3>\n"
            for viz_path in visualization_paths:
                report_string += f"<img src='{os.path.basename(viz_path)}' alt='GRC Visualization'><br>\n"

        report_string += "</body>\n</html>"

    elif output_format == 'markdown': # New markdown format
        report_string = f"""
# GRC Analysis Report
"""
        if include_summary:
            report_string += f"## Summary\n\n{summary}\n\n"

        if include_score_badge:
            report_string += f"## GRC Score & Badge\n\nGRC Score: **{grc_score}/100**\nGRC Badge: **{grc_badge}**\n\n"

        report_string += f"## Analysis Details\n\nAnalysis Recipe Used: **{recipe_used}**\n\n"

        if include_keywords:
            report_string += f"### Extracted Keywords:\n"
            if extracted_keywords:
                for kw in extracted_keywords:
                    report_string += f"- {kw}\n"
            else:
                report_string += "- None\n"
            report_string += "\n"

        if include_findings:
            report_string += f"### Findings:\n"
            if findings:
                for finding in findings:
                    report_string += f"- {finding}\n"
            else:
                report_string += "- No findings to report.\n"
            report_string += "\n"

        if include_visualizations and visualization_paths:
            report_string += "### Visualizations:\n"
            for viz_path in visualization_paths:
                report_string += f"![GRC Visualization]({os.path.basename(viz_path)})\n\n"

    else:  # Default to 'text' format
        report_string = f"""
--- GRC Analysis Report ---
"""

        if include_summary:
            report_string += f"Summary: {summary}\n\n"

        if include_score_badge:
            report_string += f"GRC Score: {grc_score}/100\n"
            report_string += f"GRC Badge: {grc_badge}\n\n"

        report_string += f"Analysis Recipe Used: {recipe_used}\n"

        if include_keywords:
            report_string += f"Extracted Keywords: {', '.join(extracted_keywords) if extracted_keywords else 'None'}\n\n"

        if include_findings:
            report_string += "Findings:\n"
            for finding in findings:
                report_string += f"- {finding}\n"

        if include_visualizations and visualization_paths:
            report_string += "\nVisualizations:\n"
            for viz_path in visualization_paths:
                report_string += f"- {viz_path}\n"

        report_string += "---------------------------"

    return report_string

# Main GRC document processor function (updated to use format_grc_report with customization options)
def main_grc_doc_processor(document_folder, analysis_recipe_name='Standard_GRC', output_file_path=None, output_format='markdown', output_visualization_dir='grc_visualizations', **report_format_options):
    print(f"Starting GRC document processing in folder: {document_folder} with recipe: {analysis_recipe_name}")
    if not os.path.exists(document_folder):
        print(f"Folder not found: {document_folder}")
        return

    for filename in os.listdir(document_folder):
        file_path = os.path.join(document_folder, filename)
        if os.path.isfile(file_path):
            print(f"\n--- Processing {filename} with recipe {analysis_recipe_name} ---")
            doc_content = read_document(file_path)
            if not doc_content.startswith("Error"):
                docling_output = process_document_for_docling(doc_content, analysis_recipe_name)
                grc_output = perform_grc_checks(docling_output, analysis_recipe_name)
                grc_score = calculate_grc_score(docling_output, analysis_recipe_name) # Calculate the score
                grc_badge = generate_grc_badge(grc_score) # Generate the badge
                grc_output['grc_score'] = grc_score # Add score to the output
                grc_output['grc_badge'] = grc_badge # Add badge to the output

                # Generate visualizations and add their paths to grc_output
                full_output_viz_dir = os.path.join(document_folder, output_visualization_dir)
                visualization_paths = generate_grc_visualizations(grc_output, full_output_viz_dir)
                grc_output['visualization_paths'] = visualization_paths

                # Format the report using the new function with custom options and output format
                formatted_report = format_grc_report(grc_output, output_format=output_format, **report_format_options)

                if output_file_path:
                    base_name, _ = os.path.splitext(output_file_path)
                    if output_format == 'html':
                        final_output_path = f"{base_name}.html"
                    elif output_format == 'markdown': # Handle markdown file extension
                        final_output_path = f"{base_name}.md"
                    else:
                        final_output_path = f"{base_name}.txt"

                    output_dir = os.path.dirname(final_output_path)
                    if output_dir and not os.path.exists(output_dir):
                        os.makedirs(output_dir, exist_ok=True)
                    with open(final_output_path, 'w', encoding='utf-8') as f:
                        f.write(formatted_report)
                    print(f"Final GRC report for {filename} saved to {final_output_path}")
                else:
                    print(f"Final GRC report for {filename}:\n{formatted_report}")
            else:
                print(doc_content)

# Example usage to demonstrate scoring and badge generation
document_directory = 'my_documents'
os.makedirs(document_directory, exist_ok=True)
# Create a dummy file for demonstration, now including 'ethics'
with open(os.path.join(document_directory, 'sample_doc.txt'), 'w') as f:
    f.write("This document discusses compliance, risk management, data protection, privacy, and cybersecurity. It also mentions ethics.")

# --- Demonstrating 'Standard_GRC' recipe with full formatted HTML output including visualizations ---
print("\n--- Demonstrating 'Standard_GRC' recipe with full formatted HTML output including visualizations ---")
output_html_file_with_viz = os.path.join(document_directory, 'standard_grc_report_with_viz.html')
main_grc_doc_processor(document_directory, analysis_recipe_name='Standard_GRC', output_file_path=output_html_file_with_viz, output_format='html', include_visualizations=True)

print(f"\nContent of {output_html_file_with_viz}: (showing first 1000 chars)\n")
if os.path.exists(output_html_file_with_viz):
    with open(output_html_file_with_viz, 'r', encoding='utf-8') as f:
        print(f.read()[:1000])
else:
    print(f"File not found: {output_html_file_with_viz}")

# --- Demonstrating 'Privacy_Focused_GRC' recipe with full formatted HTML output including visualizations ---
print("\n--- Demonstrating 'Privacy_Focused_GRC' recipe with full formatted HTML output including visualizations ---")
output_html_file_with_viz_privacy = os.path.join(document_directory, 'privacy_grc_report_with_viz.html')
main_grc_doc_processor(document_directory, analysis_recipe_name='Privacy_Focused_GRC', output_file_path=output_html_file_with_viz_privacy, output_format='html', include_visualizations=True)

print(f"\nContent of {output_html_file_with_viz_privacy}: (showing first 1000 chars)\n")
if os.path.exists(output_html_file_with_viz_privacy):
    with open(output_html_file_with_viz_privacy, 'r', encoding='utf-8') as f:
        print(f.read()[:1000])
else:
    print(f"File not found: {output_html_file_with_viz_privacy}")

# --- Demonstrating 'Standard_GRC' recipe with full formatted Markdown output including visualizations ---
print("\n--- Demonstrating 'Standard_GRC' recipe with full formatted Markdown output including visualizations ---")
output_md_file_with_viz = os.path.join(document_directory, 'standard_grc_report_with_viz.md')
main_grc_doc_processor(document_directory, analysis_recipe_name='Standard_GRC', output_file_path=output_md_file_with_viz, output_format='markdown', include_visualizations=True)

print(f"\nContent of {output_md_file_with_viz}: (showing first 1000 chars)\n")
if os.path.exists(output_md_file_with_viz):
    with open(output_md_file_with_viz, 'r', encoding='utf-8') as f:
        print(f.read()[:1000])
else:
    print(f"File not found: {output_md_file_with_viz}")
```

```python
import os
import PyPDF2 # For PDF files
import docx # For DOCX files
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.offline as py

# Function to simulate document parsing/reading
def read_document(file_path):
    print(f"Reading document: {file_path}")
    try:
        if file_path.lower().endswith('.pdf'):
            with open(file_path, 'rb') as f:
                reader = PyPDF2.PdfReader(f)
                content = "".join([page.extract_text() for page in reader.pages if page.extract_text()])
            return content
        elif file_path.lower().endswith('.docx'):
            document = docx.Document(file_path)
            content = "\n".join([para.text for para in document.paragraphs])
            return content
        else:
            # Assume plain text for other file types
            with open(file_path, 'r', encoding='utf-8') as f:
                content = f.read()
            return content
    except FileNotFoundError:
        return f"Error: Document not found at {file_path}"
    except Exception as e:
        return f"Error reading document: {e}"

# Function to simulate 'docling' processing
def process_document_for_docling(document_content, analysis_recipe_name='Standard_GRC'):
    print(f"Processing document content for 'docling' using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for keyword extraction.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    keywords = recipe_to_use['keywords']
    found_keywords = [kw for kw in keywords if kw in document_content.lower()]

    processed_data = {
        'document_summary': document_content[:100],
        'extracted_keywords': found_keywords,
        'keyword_count': len(found_keywords),
        'recipe_used': analysis_recipe_name
    }
    print(f"Extracted keywords: {found_keywords}")
    return processed_data

# Function for GRC related operations
def perform_grc_checks(processed_data, analysis_recipe_name='Standard_GRC'):
    print(f"Performing GRC checks using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for GRC checks.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    grc_categories = recipe_to_use['grc_categories']
    extracted_keywords = processed_data.get('extracted_keywords', [])

    grc_findings = []
    categorized_findings = {}
    for category, category_keywords in grc_categories.items():
        for kw in category_keywords:
            if kw in extracted_keywords:
                if category not in categorized_findings:
                    categorized_findings[category] = []
                categorized_findings[category].append(f"Keyword '{kw}' detected in {category} area.")

    if categorized_findings:
        for category, findings in categorized_findings.items():
            grc_findings.append(f"{category} Findings: {'; '.join(findings)}")
    else:
        grc_findings.append(f"No specific GRC flags based on defined categories for {analysis_recipe_name}.")

    grc_report = {
        'summary': f"GRC analysis based on '{analysis_recipe_name}' recipe output.",
        'findings': grc_findings,
        'docling_output': processed_data,
        'recipe_used_for_checks': analysis_recipe_name
    }
    return grc_report

def calculate_grc_score(processed_data, analysis_recipe_name='Standard_GRC'):
    max_score = 100
    current_score = max_score
    default_deduction = 5 # Default weight for categories not explicitly listed

    print(f"Calculating GRC score for recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found for scoring. Using 'Standard_GRC'.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    grc_categories = recipe_to_use.get('grc_categories', {})
    category_weights = recipe_to_use.get('category_weights', {})
    extracted_keywords = processed_data.get('extracted_keywords', [])

    findings_categories = set()
    for category, category_keywords in grc_categories.items():
        for kw in category_keywords:
            if kw in extracted_keywords:
                findings_categories.add(category)
                break # Only need to find one keyword to mark category as having findings

    for category_with_finding in findings_categories:
        deduction = category_weights.get(category_with_finding, default_deduction)
        print(f"  - Found findings in category '{category_with_finding}', deducting {deduction} points.")
        current_score -= deduction
        if current_score < 0:
            current_score = 0
            break

    print(f"GRC Score: {current_score}/{max_score}")
    return current_score

def generate_grc_badge(grc_score):
    """
    Generates a textual representation of a GRC badge based on the GRC score.

    Args:
        grc_score (int): The calculated GRC score.

    Returns:
        str: The badge level (e.g., 'Gold', 'Silver', 'Bronze', 'None').
    """
    if grc_score >= 90:
        return 'Gold'
    elif grc_score >= 70:
        return 'Silver'
    elif grc_score >= 50:
        return 'Bronze'
    else:
        return 'None'

def generate_grc_visualizations(grc_output_dict, output_dir):
    """
    Generates visual representations (bar chart, pie chart, interactive bar chart) from GRC output data.

    Args:
        grc_output_dict (dict): The dictionary containing GRC analysis results.
        output_dir (str): The directory to save the generated plots.

    Returns:
        list: A list of file paths to the saved visualization images/html files.
    """
    os.makedirs(output_dir, exist_ok=True)
    visualization_paths = []

    # Extract data for visualizations
    findings = grc_output_dict.get('findings', [])
    extracted_keywords = grc_output_dict.get('docling_output', {}).get('extracted_keywords', [])
    analysis_recipe_name = grc_output_dict.get('docling_output', {}).get('recipe_used', 'N/A')

    # --- Prepare data for Bar Chart (Findings per Category) ---
    findings_counts = {}
    for finding_str in findings:
        if ':' in finding_str:
            category_name = finding_str.split(':')[0].strip()
            if ' Findings' in category_name:
                category_name = category_name.replace(' Findings', '')
            findings_counts[category_name] = findings_counts.get(category_name, 0) + 1

    # Sort findings_counts by category name for consistent plotting
    sorted_findings = sorted(findings_counts.items())

    if sorted_findings:
        categories = [item[0] for item in sorted_findings]
        counts = [item[1] for item in sorted_findings]

        # Matplotlib Bar Chart
        plt.figure(figsize=(10, 6))
        plt.bar(categories, counts, color='skyblue')
        plt.xlabel('GRC Category')
        plt.ylabel('Number of Findings')
        plt.title(f'Number of GRC Findings per Category ({analysis_recipe_name})')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        bar_chart_path = os.path.join(output_dir, f'{analysis_recipe_name.lower().replace(" ", "_")}_grc_findings_bar_chart.png')
        plt.savefig(bar_chart_path)
        plt.close()
        visualization_paths.append(bar_chart_path)

        # Plotly Interactive Bar Chart
        hover_text = [f'Category: {cat}<br>Findings: {count}' for cat, count in zip(categories, counts)]
        fig = go.Figure(
            data=[go.Bar(x=categories, y=counts, hoverinfo='text', hovertext=hover_text, marker_color='lightcoral')],
            layout=go.Layout(
                title=go.layout.Title(text=f'Interactive GRC Findings per Category ({analysis_recipe_name})'),
                xaxis=go.layout.XAxis(title_text='GRC Category'),
                yaxis=go.layout.YAxis(title_text='Number of Findings')
            )
        )
        interactive_bar_chart_path = os.path.join(output_dir, f'{analysis_recipe_name.lower().replace(" ", "_")}_grc_findings_interactive_bar_chart.html')
        py.plot(fig, filename=interactive_bar_chart_path, auto_open=False)
        visualization_paths.append(interactive_bar_chart_path)

    # --- Prepare data for Pie Chart (Keyword Distribution) ---
    if extracted_keywords:
        keyword_counts = {kw: extracted_keywords.count(kw) for kw in set(extracted_keywords)}

        labels = list(keyword_counts.keys())
        sizes = list(keyword_counts.values())

        plt.figure(figsize=(10, 7))
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, textprops={'fontsize': 9})
        plt.axis('equal')
        plt.title(f'Distribution of Extracted Keywords ({analysis_recipe_name})')
        plt.tight_layout()
        pie_chart_path = os.path.join(output_dir, f'{analysis_recipe_name.lower().replace(" ", "_")}_grc_keywords_pie_chart.png')
        plt.savefig(pie_chart_path)
        plt.close()
        visualization_paths.append(pie_chart_path)

    print(f"Generated visualizations: {visualization_paths}")
    return visualization_paths

def format_grc_report(grc_output_dict,
                      include_summary=True,
                      include_score_badge=True,
                      include_keywords=True,
                      include_findings=True,
                      include_visualizations=True,
                      output_format='markdown'): # Changed default to markdown
    """
    Formats the raw GRC output dictionary into a human-readable string,
    with configurable sections and output format (text, html, or markdown).

    Args:
        grc_output_dict (dict): The dictionary containing GRC analysis results,
                                 including score and badge.
        include_summary (bool): Whether to include the summary section.
        include_score_badge (bool): Whether to include the GRC score and badge section.
        include_keywords (bool): Whether to include the extracted keywords section.
        include_findings (bool): Whether to include the detailed findings section.
        include_visualizations (bool): Whether to include links/references to visualizations.
        output_format (str): The desired output format ('text', 'html', or 'markdown').

    Returns:
        str: A multi-line string representing the formatted GRC report in the specified format.
    """
    summary = grc_output_dict.get('summary', 'No summary available.')
    findings = grc_output_dict.get('findings', ['No findings to report.'])
    grc_score = grc_output_dict.get('grc_score', 'N/A')
    grc_badge = grc_output_dict.get('grc_badge', 'N/A')
    docling_output = grc_output_dict.get('docling_output', {})
    extracted_keywords = docling_output.get('extracted_keywords', [])
    recipe_used = docling_output.get('recipe_used', 'N/A')
    visualization_paths = grc_output_dict.get('visualization_paths', [])

    if output_format == 'html':
        report_string = """<!DOCTYPE html>
<html>
<head>
<title>GRC Analysis Report</title>
<style>
body { font-family: Arial, sans-serif; line-height: 1.6; color: #333; }
h1 { color: #2c3e50; border-bottom: 2px solid #3498db; padding-bottom: 10px; }
h2 { color: #34495e; }
ul { list-style-type: disc; margin-left: 20px; }
li { margin-bottom: 5px; }
.score-badge { font-weight: bold; color: #e74c3c; }
.gold { color: #ffd700; }
.silver { color: #c0c0c0; }
.bronze { color: #cd7f32; }
img { max-width: 100%; height: auto; margin-top: 10px; border: 1px solid #ddd; }
</style>
</head>
<body>
<h1>GRC Analysis Report</h1>
"""

        if include_summary:
            report_string += f"<h2>Summary</h2><p>{summary}</p>\n"

        if include_score_badge:
            badge_class = grc_badge.lower() if grc_badge != 'N/A' and grc_badge != 'None' else ''
            report_string += f"<h2>GRC Score & Badge</h2><p>GRC Score: <strong>{grc_score}/100</strong><br>GRC Badge: <span class=\"score-badge {badge_class}\">{grc_badge}</span></p>\n"

        report_string += f"<h2>Analysis Details</h2><p>Analysis Recipe Used: <strong>{recipe_used}</strong></p>\n"

        if include_keywords:
            report_string += f"<h3>Extracted Keywords:</h3>\n<ul>\n"
            if extracted_keywords:
                for kw in extracted_keywords:
                    report_string += f"<li>{kw}</li>\n"
            else:
                report_string += "<li>None</li>\n"
            report_string += "</ul>\n"

        if include_findings:
            report_string += f"<h3>Findings:</h3>\n<ul>\n"
            if findings:
                for finding in findings:
                    report_string += f"<li>{finding}</li>\n"
            else:
                report_string += "<li>No findings to report.</li>\n"
            report_string += "</ul>\n"

        if include_visualizations and visualization_paths:
            report_string += "<h3>Visualizations:</h3>\n"
            for viz_path in visualization_paths:
                # Determine if it's an image or an HTML plot
                if viz_path.endswith('.html'):
                    report_string += f"<p><a href='{os.path.basename(viz_path)}' target='_blank'>View Interactive Plot: {os.path.basename(viz_path)}</a></p>\n"
                else:
                    report_string += f"<img src='{os.path.basename(viz_path)}' alt='GRC Visualization'><br>\n"

        report_string += "</body>\n</html>"

    elif output_format == 'markdown': # New markdown format
        report_string = f"""
# GRC Analysis Report
"""
        if include_summary:
            report_string += f"## Summary\n\n{summary}\n\n"

        if include_score_badge:
            report_string += f"## GRC Score & Badge\n\nGRC Score: **{grc_score}/100**\nGRC Badge: **{grc_badge}**\n\n"

        report_string += f"## Analysis Details\n\nAnalysis Recipe Used: **{recipe_used}**\n\n"

        if include_keywords:
            report_string += f"### Extracted Keywords:\n"
            if extracted_keywords:
                for kw in extracted_keywords:
                    report_string += f"- {kw}\n"
            else:
                report_string += "- None\n"
            report_string += "\n"

        if include_findings:
            report_string += f"### Findings:\n"
            if findings:
                for finding in findings:
                    report_string += f"- {finding}\n"
            else:
                report_string += "- No findings to report.\n"
            report_string += "\n"

        if include_visualizations and visualization_paths:
            report_string += "### Visualizations:\n"
            for viz_path in visualization_paths:
                if viz_path.endswith('.html'):
                     report_string += f"- [View Interactive Plot: {os.path.basename(viz_path)}]({os.path.basename(viz_path)})\n"
                else:
                    report_string += f"![GRC Visualization]({os.path.basename(viz_path)})\n\n"

    else:  # Default to 'text' format
        report_string = f"""
--- GRC Analysis Report ---
"""

        if include_summary:
            report_string += f"Summary: {summary}\n\n"

        if include_score_badge:
            report_string += f"GRC Score: {grc_score}/100\n"
            report_string += f"GRC Badge: {grc_badge}\n\n"

        report_string += f"Analysis Recipe Used: {recipe_used}\n"

        if include_keywords:
            report_string += f"Extracted Keywords: {', '.join(extracted_keywords) if extracted_keywords else 'None'}\n\n"

        if include_findings:
            report_string += "Findings:\n"
            for finding in findings:
                report_string += f"- {finding}\n"

        if include_visualizations and visualization_paths:
            report_string += "\nVisualizations:\n"
            for viz_path in visualization_paths:
                report_string += f"- {viz_path}\n"

        report_string += "---------------------------"

    return report_string

# Main GRC document processor function (updated to use format_grc_report with customization options)
def main_grc_doc_processor(document_folder, analysis_recipe_name='Standard_GRC', output_file_path=None, output_format='markdown', output_visualization_dir='grc_visualizations', **report_format_options):
    print(f"Starting GRC document processing in folder: {document_folder} with recipe: {analysis_recipe_name}")
    if not os.path.exists(document_folder):
        print(f"Folder not found: {document_folder}")
        return

    for filename in os.listdir(document_folder):
        file_path = os.path.join(document_folder, filename)
        if os.path.isfile(file_path):
            print(f"\n--- Processing {filename} with recipe {analysis_recipe_name} ---")
            doc_content = read_document(file_path)
            if not doc_content.startswith("Error"):
                docling_output = process_document_for_docling(doc_content, analysis_recipe_name)
                grc_output = perform_grc_checks(docling_output, analysis_recipe_name)
                grc_score = calculate_grc_score(docling_output, analysis_recipe_name) # Calculate the score
                grc_badge = generate_grc_badge(grc_score) # Generate the badge
                grc_output['grc_score'] = grc_score # Add score to the output
                grc_output['grc_badge'] = grc_badge # Add badge to the output

                # Generate visualizations and add their paths to grc_output
                full_output_viz_dir = os.path.join(document_folder, output_visualization_dir)
                visualization_paths = generate_grc_visualizations(grc_output, full_output_viz_dir)
                grc_output['visualization_paths'] = visualization_paths

                # Format the report using the new function with custom options and output format
                formatted_report = format_grc_report(grc_output, output_format=output_format, **report_format_options)

                if output_file_path:
                    base_name, _ = os.path.splitext(output_file_path)
                    if output_format == 'html':
                        final_output_path = f"{base_name}.html"
                    elif output_format == 'markdown': # Handle markdown file extension
                        final_output_path = f"{base_name}.md"
                    else:
                        final_output_path = f"{base_name}.txt"

                    output_dir = os.path.dirname(final_output_path)
                    if output_dir and not os.path.exists(output_dir):
                        os.makedirs(output_dir, exist_ok=True)
                    with open(final_output_path, 'w', encoding='utf-8') as f:
                        f.write(formatted_report)
                    print(f"Final GRC report for {filename} saved to {final_output_path}")
                else:
                    print(f"Final GRC report for {filename}:\n{formatted_report}")
            else:
                print(doc_content)

# Example usage to demonstrate scoring and badge generation
document_directory = 'my_documents'
os.makedirs(document_directory, exist_ok=True)
# Create a dummy file for demonstration, now including 'ethics'
with open(os.path.join(document_directory, 'sample_doc.txt'), 'w') as f:
    f.write("This document discusses compliance, risk management, data protection, privacy, and cybersecurity. It also mentions ethics.")

# --- Demonstrating 'Standard_GRC' recipe with full formatted HTML output including visualizations ---
print("\n--- Demonstrating 'Standard_GRC' recipe with full formatted HTML output including visualizations ---")
output_html_file_with_viz = os.path.join(document_directory, 'standard_grc_report_with_viz.html')
main_grc_doc_processor(document_directory, analysis_recipe_name='Standard_GRC', output_file_path=output_html_file_with_viz, output_format='html', include_visualizations=True)

print(f"\nContent of {output_html_file_with_viz}: (showing first 1000 chars)\n")
if os.path.exists(output_html_file_with_viz):
    with open(output_html_file_with_viz, 'r', encoding='utf-8') as f:
        print(f.read()[:1000])
else:
    print(f"File not found: {output_html_file_with_viz}")

# --- Demonstrating 'Privacy_Focused_GRC' recipe with full formatted HTML output including visualizations ---
print("\n--- Demonstrating 'Privacy_Focused_GRC' recipe with full formatted HTML output including visualizations ---")
output_html_file_with_viz_privacy = os.path.join(document_directory, 'privacy_grc_report_with_viz.html')
main_grc_doc_processor(document_directory, analysis_recipe_name='Privacy_Focused_GRC', output_file_path=output_html_file_with_viz_privacy, output_format='html', include_visualizations=True)

print(f"\nContent of {output_html_file_with_viz_privacy}: (showing first 1000 chars)\n")
if os.path.exists(output_html_file_with_viz_privacy):
    with open(output_html_file_with_viz_privacy, 'r', encoding='utf-8') as f:
        print(f.read()[:1000])
else:
    print(f"File not found: {output_html_file_with_viz_privacy}")

# --- Demonstrating 'Standard_GRC' recipe with full formatted Markdown output including visualizations ---
print("\n--- Demonstrating 'Standard_GRC' recipe with full formatted Markdown output including visualizations ---")
output_md_file_with_viz = os.path.join(document_directory, 'standard_grc_report_with_viz.md')
main_grc_doc_processor(document_directory, analysis_recipe_name='Standard_GRC', output_file_path=output_md_file_with_viz, output_format='markdown', include_visualizations=True)

print(f"\nContent of {output_md_file_with_viz}: (showing first 1000 chars)\n")
if os.path.exists(output_md_file_with_viz):
    with open(output_md_file_with_viz, 'r', encoding='utf-8') as f:
        print(f.read()[:1000])
else:
    print(f"File not found: {output_md_file_with_viz}")
```

```python
import os
import PyPDF2 # For PDF files
import docx # For DOCX files
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.offline as py
from datetime import datetime, timedelta

# Function to simulate document parsing/reading
def read_document(file_path):
    print(f"Reading document: {file_path}")
    try:
        if file_path.lower().endswith('.pdf'):
            with open(file_path, 'rb') as f:
                reader = PyPDF2.PdfReader(f)
                content = "".join([page.extract_text() for page in reader.pages if page.extract_text()])
            return content
        elif file_path.lower().endswith('.docx'):
            document = docx.Document(file_path)
            content = "\n".join([para.text for para in document.paragraphs])
            return content
        else:
            # Assume plain text for other file types
            with open(file_path, 'r', encoding='utf-8') as f:
                content = f.read()
            return content
    except FileNotFoundError:
        return f"Error: Document not found at {file_path}"
    except Exception as e:
        return f"Error reading document: {e}"

# Function to simulate 'docling' processing
def process_document_for_docling(document_content, analysis_recipe_name='Standard_GRC'):
    print(f"Processing document content for 'docling' using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for keyword extraction.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    keywords = recipe_to_use['keywords']
    found_keywords = [kw for kw in keywords if kw in document_content.lower()]

    processed_data = {
        'document_summary': document_content[:100],
        'extracted_keywords': found_keywords,
        'keyword_count': len(found_keywords),
        'recipe_used': analysis_recipe_name
    }
    print(f"Extracted keywords: {found_keywords}")
    return processed_data

# Function for GRC related operations
def perform_grc_checks(processed_data, analysis_recipe_name='Standard_GRC'):
    print(f"Performing GRC checks using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for GRC checks.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    grc_categories = recipe_to_use['grc_categories']
    extracted_keywords = processed_data.get('extracted_keywords', [])

    grc_findings = []
    categorized_findings = {}
    for category, category_keywords in grc_categories.items():
        for kw in category_keywords:
            if kw in extracted_keywords:
                if category not in categorized_findings:
                    categorized_findings[category] = []
                categorized_findings[category].append(f"Keyword '{kw}' detected in {category} area.")

    if categorized_findings:
        for category, findings in categorized_findings.items():
            grc_findings.append(f"{category} Findings: {'; '.join(findings)}")
    else:
        grc_findings.append(f"No specific GRC flags based on defined categories for {analysis_recipe_name}.")

    grc_report = {
        'summary': f"GRC analysis based on '{analysis_recipe_name}' recipe output.",
        'findings': grc_findings,
        'docling_output': processed_data,
        'recipe_used_for_checks': analysis_recipe_name
    }
    return grc_report

def calculate_grc_score(processed_data, analysis_recipe_name='Standard_GRC'):
    max_score = 100
    current_score = max_score
    default_deduction = 5 # Default weight for categories not explicitly listed

    print(f"Calculating GRC score for recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found for scoring. Using 'Standard_GRC'.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    grc_categories = recipe_to_use.get('grc_categories', {})
    category_weights = recipe_to_use.get('category_weights', {})
    extracted_keywords = processed_data.get('extracted_keywords', [])

    findings_categories = set()
    for category, category_keywords in grc_categories.items():
        for kw in category_keywords:
            if kw in extracted_keywords:
                findings_categories.add(category)
                break # Only need to find one keyword to mark category as having findings

    for category_with_finding in findings_categories:
        deduction = category_weights.get(category_with_finding, default_deduction)
        print(f"  - Found findings in category '{category_with_finding}', deducting {deduction} points.")
        current_score -= deduction
        if current_score < 0:
            current_score = 0
            break

    print(f"GRC Score: {current_score}/{max_score}")
    return current_score

def generate_grc_badge(grc_score):
    """
    Generates a textual representation of a GRC badge based on the GRC score.

    Args:
        grc_score (int): The calculated GRC score.

    Returns:
        str: The badge level (e.g., 'Gold', 'Silver', 'Bronze', 'None').
    """
    if grc_score >= 90:
        return 'Gold'
    elif grc_score >= 70:
        return 'Silver'
    elif grc_score >= 50:
        return 'Bronze'
    else:
        return 'None'

def generate_grc_trend_chart(output_dir):
    """
    Generates a trend line chart of conceptual GRC scores over time.

    Args:
        output_dir (str): The directory to save the generated plot.

    Returns:
        str: The file path to the saved trend line chart image, or None if not generated.
    """
    os.makedirs(output_dir, exist_ok=True)

    # Simulate historical GRC score data (e.g., last 6 months)
    end_date = datetime.now()
    dates = [end_date - timedelta(days=30 * i) for i in range(6, -1, -1)] # 7 data points for 6 months
    # Dummy scores, with a conceptual 'trend' or fluctuation
    scores = [70, 75, 80, 78, 85, 82, 88]

    if not dates or not scores:
        print("No data available to generate GRC trend chart.")
        return None

    plt.figure(figsize=(10, 6))
    plt.plot(dates, scores, marker='o', linestyle='-', color='green')
    plt.xlabel('Date')
    plt.ylabel('GRC Score')
    plt.title('Conceptual GRC Score Trend Over Time')
    plt.grid(True)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    trend_chart_path = os.path.join(output_dir, 'grc_score_trend_chart.png')
    plt.savefig(trend_chart_path)
    plt.close()
    print(f"Generated GRC trend chart: {trend_chart_path}")
    return trend_chart_path

def generate_grc_visualizations(grc_output_dict, output_dir):
    """
    Generates visual representations (bar chart, pie chart, interactive bar chart) from GRC output data.

    Args:
        grc_output_dict (dict): The dictionary containing GRC analysis results.
        output_dir (str): The directory to save the generated plots.

    Returns:
        list: A list of file paths to the saved visualization images/html files.
    """
    os.makedirs(output_dir, exist_ok=True)
    visualization_paths = []

    # Extract data for visualizations
    findings = grc_output_dict.get('findings', [])
    extracted_keywords = grc_output_dict.get('docling_output', {}).get('extracted_keywords', [])
    analysis_recipe_name = grc_output_dict.get('docling_output', {}).get('recipe_used', 'N/A')

    # --- Prepare data for Bar Chart (Findings per Category) ---
    findings_counts = {}
    for finding_str in findings:
        if ':' in finding_str:
            category_name = finding_str.split(':')[0].strip()
            if ' Findings' in category_name:
                category_name = category_name.replace(' Findings', '')
            findings_counts[category_name] = findings_counts.get(category_name, 0) + 1

    # Sort findings_counts by category name for consistent plotting
    sorted_findings = sorted(findings_counts.items())

    if sorted_findings:
        categories = [item[0] for item in sorted_findings]
        counts = [item[1] for item in sorted_findings]

        # Matplotlib Bar Chart
        plt.figure(figsize=(10, 6))
        plt.bar(categories, counts, color='skyblue')
        plt.xlabel('GRC Category')
        plt.ylabel('Number of Findings')
        plt.title(f'Number of GRC Findings per Category ({analysis_recipe_name})')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        bar_chart_path = os.path.join(output_dir, f'{analysis_recipe_name.lower().replace(" ", "_")}_grc_findings_bar_chart.png')
        plt.savefig(bar_chart_path)
        plt.close()
        visualization_paths.append(bar_chart_path)

        # Plotly Interactive Bar Chart
        hover_text = [f'Category: {cat}<br>Findings: {count}' for cat, count in zip(categories, counts)]
        fig = go.Figure(
            data=[go.Bar(x=categories, y=counts, hoverinfo='text', hovertext=hover_text, marker_color='lightcoral')],
            layout=go.Layout(
                title=go.layout.Title(text=f'Interactive GRC Findings per Category ({analysis_recipe_name})'),
                xaxis=go.layout.XAxis(title_text='GRC Category'),
                yaxis=go.layout.YAxis(title_text='Number of Findings')
            )
        )
        interactive_bar_chart_path = os.path.join(output_dir, f'{analysis_recipe_name.lower().replace(" ", "_")}_grc_findings_interactive_bar_chart.html')
        py.plot(fig, filename=interactive_bar_chart_path, auto_open=False)
        visualization_paths.append(interactive_bar_chart_path)

    # --- Prepare data for Pie Chart (Keyword Distribution) ---
    if extracted_keywords:
        keyword_counts = {kw: extracted_keywords.count(kw) for kw in set(extracted_keywords)}

        labels = list(keyword_counts.keys())
        sizes = list(keyword_counts.values())

        plt.figure(figsize=(10, 7))
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, textprops={'fontsize': 9})
        plt.axis('equal')
        plt.title(f'Distribution of Extracted Keywords ({analysis_recipe_name})')
        plt.tight_layout()
        pie_chart_path = os.path.join(output_dir, f'{analysis_recipe_name.lower().replace(" ", "_")}_grc_keywords_pie_chart.png')
        plt.savefig(pie_chart_path)
        plt.close()
        visualization_paths.append(pie_chart_path)

    # --- Generate GRC Trend Chart (New addition) ---
    trend_chart_path = generate_grc_trend_chart(output_dir)
    if trend_chart_path:
        visualization_paths.append(trend_chart_path)

    print(f"Generated visualizations: {visualization_paths}")
    return visualization_paths

def format_grc_report(grc_output_dict,
                      include_summary=True,
                      include_score_badge=True,
                      include_keywords=True,
                      include_findings=True,
                      include_visualizations=True,
                      output_format='markdown'): # Changed default to markdown
    """
    Formats the raw GRC output dictionary into a human-readable string,
    with configurable sections and output format (text, html, or markdown).

    Args:
        grc_output_dict (dict): The dictionary containing GRC analysis results,
                                 including score and badge.
        include_summary (bool): Whether to include the summary section.
        include_score_badge (bool): Whether to include the GRC score and badge section.
        include_keywords (bool): Whether to include the extracted keywords section.
        include_findings (bool): Whether to include the detailed findings section.
        include_visualizations (bool): Whether to include links/references to visualizations.
        output_format (str): The desired output format ('text', 'html', or 'markdown').

    Returns:
        str: A multi-line string representing the formatted GRC report in the specified format.
    """
    summary = grc_output_dict.get('summary', 'No summary available.')
    findings = grc_output_dict.get('findings', ['No findings to report.'])
    grc_score = grc_output_dict.get('grc_score', 'N/A')
    grc_badge = grc_output_dict.get('grc_badge', 'N/A')
    docling_output = grc_output_dict.get('docling_output', {})
    extracted_keywords = docling_output.get('extracted_keywords', [])
    recipe_used = docling_output.get('recipe_used', 'N/A')
    visualization_paths = grc_output_dict.get('visualization_paths', [])

    if output_format == 'html':
        report_string = """<!DOCTYPE html>
<html>
<head>
<title>GRC Analysis Report</title>
<style>
body { font-family: Arial, sans-serif; line-height: 1.6; color: #333; }
h1 { color: #2c3e50; border-bottom: 2px solid #3498db; padding-bottom: 10px; }
h2 { color: #34495e; }
ul { list-style-type: disc; margin-left: 20px; }
li { margin-bottom: 5px; }
.score-badge { font-weight: bold; color: #e74c3c; }
.gold { color: #ffd700; }
.silver { color: #c0c0c0; }
.bronze { color: #cd7f32; }
img { max-width: 100%; height: auto; margin-top: 10px; border: 1px solid #ddd; }
</style>
</head>
<body>
<h1>GRC Analysis Report</h1>
"""

        if include_summary:
            report_string += f"<h2>Summary</h2><p>{summary}</p>\n"

        if include_score_badge:
            badge_class = grc_badge.lower() if grc_badge != 'N/A' and grc_badge != 'None' else ''
            report_string += f"<h2>GRC Score & Badge</h2><p>GRC Score: <strong>{grc_score}/100</strong><br>GRC Badge: <span class=\"score-badge {badge_class}\">{grc_badge}</span></p>\n"

        report_string += f"<h2>Analysis Details</h2><p>Analysis Recipe Used: <strong>{recipe_used}</strong></p>\n"

        if include_keywords:
            report_string += f"<h3>Extracted Keywords:</h3>\n<ul>\n"
            if extracted_keywords:
                for kw in extracted_keywords:
                    report_string += f"<li>{kw}</li>\n"
            else:
                report_string += "<li>None</li>\n"
            report_string += "</ul>\n"

        if include_findings:
            report_string += f"<h3>Findings:</h3>\n<ul>\n"
            if findings:
                for finding in findings:
                    report_string += f"<li>{finding}</li>\n"
            else:
                report_string += "<li>No findings to report.</li>\n"
            report_string += "</ul>\n"

        if include_visualizations and visualization_paths:
            report_string += "<h3>Visualizations:</h3>\n"
            for viz_path in visualization_paths:
                # Determine if it's an image or an HTML plot
                if viz_path.endswith('.html'):
                    report_string += f"<p><a href='{os.path.basename(viz_path)}' target='_blank'>View Interactive Plot: {os.path.basename(viz_path)}</a></p>\n"
                else:
                    report_string += f"<img src='{os.path.basename(viz_path)}' alt='GRC Visualization'><br>\n"

        report_string += "</body>\n</html>"

    elif output_format == 'markdown': # New markdown format
        report_string = f"""
# GRC Analysis Report
"""
        if include_summary:
            report_string += f"## Summary\n\n{summary}\n\n"

        if include_score_badge:
            report_string += f"## GRC Score & Badge\n\nGRC Score: **{grc_score}/100**\nGRC Badge: **{grc_badge}**\n\n"

        report_string += f"## Analysis Details\n\nAnalysis Recipe Used: **{recipe_used}**\n\n"

        if include_keywords:
            report_string += f"### Extracted Keywords:\n"
            if extracted_keywords:
                for kw in extracted_keywords:
                    report_string += f"- {kw}\n"
            else:
                report_string += "- None\n"
            report_string += "\n"

        if include_findings:
            report_string += f"### Findings:\n"
            if findings:
                for finding in findings:
                    report_string += f"- {finding}\n"
            else:
                report_string += "- No findings to report.\n"
            report_string += "\n"

        if include_visualizations and visualization_paths:
            report_string += "### Visualizations:\n"
            for viz_path in visualization_paths:
                if viz_path.endswith('.html'):
                     report_string += f"- [View Interactive Plot: {os.path.basename(viz_path)}]({os.path.basename(viz_path)})\n"
                else:
                    report_string += f"![GRC Visualization]({os.path.basename(viz_path)})\n\n"

    else:  # Default to 'text' format
        report_string = f"""
--- GRC Analysis Report ---
"""

        if include_summary:
            report_string += f"Summary: {summary}\n\n"

        if include_score_badge:
            report_string += f"GRC Score: {grc_score}/100\n"
            report_string += f"GRC Badge: {grc_badge}\n\n"

        report_string += f"Analysis Recipe Used: {recipe_used}\n"

        if include_keywords:
            report_string += f"Extracted Keywords: {', '.join(extracted_keywords) if extracted_keywords else 'None'}\n\n"

        if include_findings:
            report_string += "Findings:\n"
            for finding in findings:
                report_string += f"- {finding}\n"

        if include_visualizations and visualization_paths:
            report_string += "\nVisualizations:\n"
            for viz_path in visualization_paths:
                report_string += f"- {viz_path}\n"

        report_string += "---------------------------"

    return report_string

# Main GRC document processor function (updated to use format_grc_report with customization options)
def main_grc_doc_processor(document_folder, analysis_recipe_name='Standard_GRC', output_file_path=None, output_format='markdown', output_visualization_dir='grc_visualizations', **report_format_options):
    print(f"Starting GRC document processing in folder: {document_folder} with recipe: {analysis_recipe_name}")
    if not os.path.exists(document_folder):
        print(f"Folder not found: {document_folder}")
        return

    for filename in os.listdir(document_folder):
        file_path = os.path.join(document_folder, filename)
        if os.path.isfile(file_path):
            print(f"\n--- Processing {filename} with recipe {analysis_recipe_name} ---")
            doc_content = read_document(file_path)
            if not doc_content.startswith("Error"):
                docling_output = process_document_for_docling(doc_content, analysis_recipe_name)
                grc_output = perform_grc_checks(docling_output, analysis_recipe_name)
                grc_score = calculate_grc_score(docling_output, analysis_recipe_name) # Calculate the score
                grc_badge = generate_grc_badge(grc_score) # Generate the badge
                grc_output['grc_score'] = grc_score # Add score to the output
                grc_output['grc_badge'] = grc_badge # Add badge to the output

                # Generate visualizations and add their paths to grc_output
                full_output_viz_dir = os.path.join(document_folder, output_visualization_dir)
                visualization_paths = generate_grc_visualizations(grc_output, full_output_viz_dir)
                grc_output['visualization_paths'] = visualization_paths

                # Format the report using the new function with custom options and output format
                formatted_report = format_grc_report(grc_output, output_format=output_format, **report_format_options)

                if output_file_path:
                    base_name, _ = os.path.splitext(output_file_path)
                    if output_format == 'html':
                        final_output_path = f"{base_name}.html"
                    elif output_format == 'markdown': # Handle markdown file extension
                        final_output_path = f"{base_name}.md"
                    else:
                        final_output_path = f"{base_name}.txt"

                    output_dir = os.path.dirname(final_output_path)
                    if output_dir and not os.path.exists(output_dir):
                        os.makedirs(output_dir, exist_ok=True)
                    with open(final_output_path, 'w', encoding='utf-8') as f:
                        f.write(formatted_report)
                    print(f"Final GRC report for {filename} saved to {final_output_path}")
                else:
                    print(f"Final GRC report for {filename}:\n{formatted_report}")
            else:
                print(doc_content)

# Example usage to demonstrate scoring and badge generation
document_directory = 'my_documents'
os.makedirs(document_directory, exist_ok=True)
# Create a dummy file for demonstration, now including 'ethics'
with open(os.path.join(document_directory, 'sample_doc.txt'), 'w') as f:
    f.write("This document discusses compliance, risk management, data protection, privacy, and cybersecurity. It also mentions ethics.")

# --- Demonstrating 'Standard_GRC' recipe with full formatted HTML output including visualizations ---
print("\n--- Demonstrating 'Standard_GRC' recipe with full formatted HTML output including visualizations ---")
output_html_file_with_viz = os.path.join(document_directory, 'standard_grc_report_with_viz.html')
main_grc_doc_processor(document_directory, analysis_recipe_name='Standard_GRC', output_file_path=output_html_file_with_viz, output_format='html', include_visualizations=True)

print(f"\nContent of {output_html_file_with_viz}: (showing first 1000 chars)\n")
if os.path.exists(output_html_file_with_viz):
    with open(output_html_file_with_viz, 'r', encoding='utf-8') as f:
        print(f.read()[:1000])
else:
    print(f"File not found: {output_html_file_with_viz}")

# --- Demonstrating 'Privacy_Focused_GRC' recipe with full formatted HTML output including visualizations ---
print("\n--- Demonstrating 'Privacy_Focused_GRC' recipe with full formatted HTML output including visualizations ---")
output_html_file_with_viz_privacy = os.path.join(document_directory, 'privacy_grc_report_with_viz.html')
main_grc_doc_processor(document_directory, analysis_recipe_name='Privacy_Focused_GRC', output_file_path=output_html_file_with_viz_privacy, output_format='html', include_visualizations=True)

print(f"\nContent of {output_html_file_with_viz_privacy}: (showing first 1000 chars)\n")
if os.path.exists(output_html_file_with_viz_privacy):
    with open(output_html_file_with_viz_privacy, 'r', encoding='utf-8') as f:
        print(f.read()[:1000])
else:
    print(f"File not found: {output_html_file_with_viz_privacy}")

# --- Demonstrating 'Standard_GRC' recipe with full formatted Markdown output including visualizations ---
print("\n--- Demonstrating 'Standard_GRC' recipe with full formatted Markdown output including visualizations ---")
output_md_file_with_viz = os.path.join(document_directory, 'standard_grc_report_with_viz.md')
main_grc_doc_processor(document_directory, analysis_recipe_name='Standard_GRC', output_file_path=output_md_file_with_viz, output_format='markdown', include_visualizations=True)

print(f"\nContent of {output_md_file_with_viz}: (showing first 1000 chars)\n")
if os.path.exists(output_md_file_with_viz):
    with open(output_md_file_with_viz, 'r', encoding='utf-8') as f:
        print(f.read()[:1000])
else:
    print(f"File not found: {output_md_file_with_viz}")
```

```python
import os
import PyPDF2 # For PDF files
import docx # For DOCX files
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.offline as py
from datetime import datetime, timedelta

# --- ANALYSIS_RECIPES: Consolidated from all previous modifications ---
ANALYSIS_RECIPES = {
    'Standard_GRC': {
        'keywords': ['compliance', 'regulation', 'risk', 'audit', 'policy', 'governance', 'security', 'privacy', 'ethics'],
        'grc_categories': {
            'Compliance': ['compliance', 'regulation', 'policy'],
            'Risk Management': ['risk', 'security', 'privacy'],
            'Governance': ['governance', 'audit'],
            'Ethics': ['ethics']
        },
        'category_weights': {
            'Compliance': 5,
            'Risk Management': 10,
            'Governance': 3,
            'Ethics': 7
        }
    },
    'Privacy_Focused_GRC': {
        'keywords': ['privacy', 'data protection', 'GDPR', 'CCPA', 'PII', 'consent', 'security', 'compliance'],
        'grc_categories': {
            'Data Privacy': ['privacy', 'data protection', 'GDPR', 'CCPA', 'PII', 'consent'],
            'Security Compliance': ['security', 'compliance']
        },
        'category_weights': {
            'Data Privacy': 8,
            'Security Compliance': 6
        }
    },
    'Security_Focused_GRC': {
        'keywords': ['security', 'cybersecurity', 'breach', 'vulnerability', 'encryption', 'firewall', 'incident', 'data loss'],
        'grc_categories': {
            'Data Security': ['security', 'cybersecurity', 'encryption', 'firewall'],
            'Incident Response': ['breach', 'vulnerability', 'incident', 'data loss']
        },
        'category_weights': {
            'Data Security': 7,
            'Incident Response': 10
        }
    },
    'Legal_Compliance_GRC': {
        'keywords': ['legal', 'law', 'regulation', 'contract', 'litigation', 'fine', 'penalty', 'jurisdiction'],
        'grc_categories': {
            'Legal Compliance': ['legal', 'law', 'regulation', 'contract'],
            'Legal Risk': ['litigation', 'fine', 'penalty', 'jurisdiction']
        },
        'category_weights': {
            'Legal Compliance': 6,
            'Legal Risk': 12
        }
    },
    'OEI_Organisational_Ethics_Indicator': {
        'keywords': ['ethics', 'integrity', 'values', 'conduct', 'culture', 'governance', 'transparency', 'accountability', 'compliance', 'misconduct', 'whistleblower', 'dignity', 'respect', 'fairness'],
        'grc_categories': {
            'Ethical Culture': ['ethics', 'integrity', 'values', 'culture', 'respect', 'dignity', 'fairness'],
            'Conduct and Accountability': ['conduct', 'transparency', 'accountability', 'governance', 'misconduct'],
            'Whistleblower Protection': ['whistleblower', 'reporting', 'retaliation'],
            'Ethics Compliance': ['compliance', 'code of conduct']
        },
        'category_weights': {
            'Ethical Culture': 7,
            'Conduct and Accountability': 9,
            'Whistleblower Protection': 10,
            'Ethics Compliance': 8
        }
    },
    'Framework_Gap_Analysis_GRC': {
        'keywords': ['framework', 'standard', 'compliance', 'gap', 'assessment', 'alignment', 'recommendation', 'control', 'policy', 'procedure', 'best practice', 'regulatory', 'deficiency'],
        'grc_categories': {
            'Framework Alignment': ['framework', 'standard', 'alignment', 'best practice'],
            'Gap Identification': ['gap', 'assessment', 'deficiency'],
            'Remediation Planning': ['recommendation', 'control', 'policy', 'procedure'],
            'Regulatory Compliance': ['compliance', 'regulatory']
        },
        'category_weights': {
            'Framework Alignment': 5,
            'Gap Identification': 9,
            'Remediation Planning': 7,
            'Regulatory Compliance': 8
        }
    },
    'Vendor_Risk_Management_GRC': {
        'keywords': ['vendor', 'third-party', 'contract', 'SLA', 'service level', 'due diligence', 'onboarding', 'offboarding', 'data sharing', 'supply chain', 'resilience', 'security breach'],
        'grc_categories': {
            'Vendor Due Diligence': ['vendor', 'third-party', 'due diligence', 'onboarding'],
            'Contractual Compliance': ['contract', 'SLA', 'service level', 'data sharing'],
            'Supply Chain Risk': ['supply chain', 'resilience', 'security breach']
        },
        'category_weights': {
            'Vendor Due Diligence': 6,
            'Contractual Compliance': 8,
            'Supply Chain Risk': 11
        }
    },
    'ESG_Sustainability_GRC': {
        'keywords': ['sustainability', 'environmental', 'social', 'governance', 'carbon footprint', 'emissions', 'diversity', 'inclusion', 'community', 'ethics', 'CSR', 'reporting', 'impact', 'stakeholder'],
        'grc_categories': {
            'Environmental Impact': ['sustainability', 'environmental', 'carbon footprint', 'emissions'],
            'Social Responsibility': ['social', 'diversity', 'inclusion', 'community', 'CSR', 'stakeholder'],
            'Governance & Ethics (ESG)': ['governance', 'ethics', 'reporting', 'impact']
        },
        'category_weights': {
            'Environmental Impact': 7,
            'Social Responsibility': 6,
            'Governance & Ethics (ESG)': 8
        }
    },
    'Corporate_Governance_Compliance_GRC': {
        'keywords': ['corporate governance', 'board of directors', 'shareholder rights', 'executive compensation', 'internal controls', 'regulatory reporting', 'fiduciary duty', 'disclosure', 'ethics committee'],
        'grc_categories': {
            'Board Oversight': ['board of directors', 'fiduciary duty', 'governance'],
            'Executive Accountability': ['executive compensation', 'executive conduct'],
            'Transparency & Disclosure': ['shareholder rights', 'regulatory reporting', 'disclosure'],
            'Internal Controls & Ethics': ['internal controls', 'ethics committee']
        },
        'category_weights': {
            'Board Oversight': 9,
            'Executive Accountability': 8,
            'Transparency & Disclosure': 7,
            'Internal Controls & Ethics': 6
        }
    },
    'Reputational_Risk_GRC': {
        'keywords': ['reputation', 'brand image', 'public trust', 'stakeholder perception', 'media coverage', 'ethical lapse', 'controversy', 'scandal', 'integrity', 'social media', 'consumer sentiment'],
        'grc_categories': {
            'Public Perception & Trust': ['reputation', 'brand image', 'public trust', 'stakeholder perception', 'consumer sentiment'],
            'Media & Communication Risk': ['media coverage', 'social media', 'controversy'],
            'Ethical Conduct Impact': ['ethical lapse', 'scandal', 'integrity']
        },
        'category_weights': {
            'Public Perception & Trust': 10,
            'Media & Communication Risk': 9,
            'Ethical Conduct Impact': 12
        }
    },
    'Legal_Jurisprudence_GRC': {
        'keywords': [
            'jurisprudence', 'precedent', 'common law', 'statutory law', 'case law',
            'constitutional law', 'judicial review', 'legal theory', 'court decision', 'legal interpretation'
        ],
        'grc_categories': {
            'Legal Theory': ['jurisprudence', 'legal theory', 'legal interpretation'],
            'Judicial Precedent': ['precedent', 'common law', 'case law', 'court decision'],
            'Constitutional Aspects': ['constitutional law', 'judicial review', 'statutory law']
        },
        'category_weights': {
            'Legal Theory': 7,
            'Judicial Precedent': 9,
            'Constitutional Aspects': 10
        }
    },
    'Cultural_Context_GRC': {
        'keywords': [
            'culture', 'organizational values', 'ethics', 'diversity', 'inclusion',
            'employee engagement', 'leadership', 'communication', 'trust',
            'behavior', 'norms', 'tradition', 'local custom', 'social fabric'
        ],
        'grc_categories': {
            'Organizational Culture': ['culture', 'organizational values', 'employee engagement', 'leadership', 'trust'],
            'Ethical & Social Norms': ['ethics', 'diversity', 'inclusion', 'behavior', 'norms'],
            'Local & Traditional Impact': ['tradition', 'local custom', 'social fabric', 'communication']
        },
        'category_weights': {
            'Organizational Culture': 8,
            'Ethical & Social Norms': 9,
            'Local & Traditional Impact': 7
        }
    },
    'Economic_Impact_GRC': {
        'keywords': [
            'economic impact', 'financial risk', 'market volatility', 'revenue', 'profit',
            'investment', 'shareholder value', 'cost reduction', 'sustainability', 'growth',
            'inflation', 'recession', 'supply chain disruption'
        ],
        'grc_categories': {
            'Financial Performance': ['economic impact', 'revenue', 'profit', 'shareholder value', 'growth'],
            'Market & Investment Risk': ['financial risk', 'market volatility', 'investment', 'inflation', 'recession'],
            'Operational Economics': ['cost reduction', 'supply chain disruption']
        },
        'category_weights': {
            'Financial Performance': 9,
            'Market & Investment Risk': 11,
            'Operational Economics': 8
        }
    },
    'Social_Impact_GRC': {
        'keywords': [
            'social impact', 'community engagement', 'human rights', 'labor practices',
            'diversity', 'inclusion', 'stakeholder relations', 'philanthropy',
            'public health', 'education access', 'social equity', 'ethical sourcing'
        ],
        'grc_categories': {
            'Community & Stakeholder Impact': ['social impact', 'community engagement', 'stakeholder relations', 'philanthropy'],
            'Human Capital & Rights': ['human rights', 'labor practices', 'diversity', 'inclusion', 'ethical sourcing'],
            'Societal Welfare': ['public health', 'education access', 'social equity']
        },
        'category_weights': {
            'Community & Stakeholder Impact': 8,
            'Human Capital & Rights': 10,
            'Societal Welfare': 7
        }
    }
}


# Function to simulate document parsing/reading
def read_document(file_path):
    print(f"Reading document: {file_path}")
    try:
        if file_path.lower().endswith('.pdf'):
            with open(file_path, 'rb') as f:
                reader = PyPDF2.PdfReader(f)
                content = "".join([page.extract_text() for page in reader.pages if page.extract_text()])
            return content
        elif file_path.lower().endswith('.docx'):
            document = docx.Document(file_path)
            content = "\n".join([para.text for para in document.paragraphs])
            return content
        else:
            # Assume plain text for other file types
            with open(file_path, 'r', encoding='utf-8') as f:
                content = f.read()
            return content
    except FileNotFoundError:
        return f"Error: Document not found at {file_path}"
    except Exception as e:
        return f"Error reading document: {e}"

# Function to simulate 'docling' processing
def process_document_for_docling(document_content, analysis_recipe_name='Standard_GRC'):
    print(f"Processing document content for 'docling' using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for keyword extraction.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    keywords = recipe_to_use['keywords']
    found_keywords = [kw for kw in keywords if kw in document_content.lower()]

    processed_data = {
        'document_summary': document_content[:100],
        'extracted_keywords': found_keywords,
        'keyword_count': len(found_keywords),
        'recipe_used': analysis_recipe_name
    }
    print(f"Extracted keywords: {found_keywords}")
    return processed_data

# Function for GRC related operations
def perform_grc_checks(processed_data, analysis_recipe_name='Standard_GRC'):
    print(f"Performing GRC checks using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for GRC checks.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    grc_categories = recipe_to_use['grc_categories']
    extracted_keywords = processed_data.get('extracted_keywords', [])

    grc_findings = []
    categorized_findings = {}
    for category, category_keywords in grc_categories.items():
        for kw in category_keywords:
            if kw in extracted_keywords:
                if category not in categorized_findings:
                    categorized_findings[category] = []
                categorized_findings[category].append(f"Keyword '{kw}' detected in {category} area.")

    if categorized_findings:
        for category, findings in categorized_findings.items():
            grc_findings.append(f"{category} Findings: {'; '.join(findings)}")
    else:
        grc_findings.append(f"No specific GRC flags based on defined categories for {analysis_recipe_name}.")

    grc_report = {
        'summary': f"GRC analysis based on '{analysis_recipe_name}' recipe output.",
        'findings': grc_findings,
        'docling_output': processed_data,
        'recipe_used_for_checks': analysis_recipe_name
    }
    return grc_report

def calculate_grc_score(processed_data, analysis_recipe_name='Standard_GRC'):
    max_score = 100
    current_score = max_score
    default_deduction = 5 # Default weight for categories not explicitly listed

    print(f"Calculating GRC score for recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found for scoring. Using 'Standard_GRC'.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    grc_categories = recipe_to_use.get('grc_categories', {})
    category_weights = recipe_to_use.get('category_weights', {})
    extracted_keywords = processed_data.get('extracted_keywords', [])

    findings_categories = set()
    for category, category_keywords in grc_categories.items():
        for kw in category_keywords:
            if kw in extracted_keywords:
                findings_categories.add(category)
                break # Only need to find one keyword to mark category as having findings

    for category_with_finding in findings_categories:
        deduction = category_weights.get(category_with_finding, default_deduction)
        print(f"  - Found findings in category '{category_with_finding}', deducting {deduction} points.")
        current_score -= deduction
        if current_score < 0:
            current_score = 0
            break

    print(f"GRC Score: {current_score}/{max_score}")
    return current_score

def generate_grc_badge(grc_score):
    """
    Generates a textual representation of a GRC badge based on the GRC score.

    Args:
        grc_score (int): The calculated GRC score.

    Returns:
        str: The badge level (e.g., 'Gold', 'Silver', 'Bronze', 'None').
    """
    if grc_score >= 90:
        return 'Gold'
    elif grc_score >= 70:
        return 'Silver'
    elif grc_score >= 50:
        return 'Bronze'
    else:
        return 'None'

def generate_grc_trend_chart(output_dir):
    """
    Generates a trend line chart of conceptual GRC scores over time.

    Args:
        output_dir (str): The directory to save the generated plot.

    Returns:
        str: The file path to the saved trend line chart image, or None if not generated.
    """
    os.makedirs(output_dir, exist_ok=True)

    # Simulate historical GRC score data (e.g., last 6 months)
    end_date = datetime.now()
    dates = [end_date - timedelta(days=30 * i) for i in range(6, -1, -1)] # 7 data points for 6 months
    # Dummy scores, with a conceptual 'trend' or fluctuation
    scores = [70, 75, 80, 78, 85, 82, 88]

    if not dates or not scores:
        print("No data available to generate GRC trend chart.")
        return None

    plt.figure(figsize=(10, 6))
    plt.plot(dates, scores, marker='o', linestyle='-', color='green')
    plt.xlabel('Date')
    plt.ylabel('GRC Score')
    plt.title('Conceptual GRC Score Trend Over Time')
    plt.grid(True)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    trend_chart_path = os.path.join(output_dir, 'grc_score_trend_chart.png')
    plt.savefig(trend_chart_path)
    plt.close()
    print(f"Generated GRC trend chart: {trend_chart_path}")
    return trend_chart_path

def generate_grc_visualizations(grc_output_dict, output_dir):
    """
    Generates visual representations (bar chart, pie chart, interactive bar chart) from GRC output data.

    Args:
        grc_output_dict (dict): The dictionary containing GRC analysis results.
        output_dir (str): The directory to save the generated plots.

    Returns:
        list: A list of file paths to the saved visualization images/html files.
    """
    os.makedirs(output_dir, exist_ok=True)
    visualization_paths = []

    # Extract data for visualizations
    findings = grc_output_dict.get('findings', [])
    extracted_keywords = grc_output_dict.get('docling_output', {}).get('extracted_keywords', [])
    analysis_recipe_name = grc_output_dict.get('docling_output', {}).get('recipe_used', 'N/A')

    # --- Prepare data for Bar Chart (Findings per Category) ---
    findings_counts = {}
    for finding_str in findings:
        if ':' in finding_str:
            category_name = finding_str.split(':')[0].strip()
            if ' Findings' in category_name:
                category_name = category_name.replace(' Findings', '')
            findings_counts[category_name] = findings_counts.get(category_name, 0) + 1

    # Sort findings_counts by category name for consistent plotting
    sorted_findings = sorted(findings_counts.items())

    if sorted_findings:
        categories = [item[0] for item in sorted_findings]
        counts = [item[1] for item in sorted_findings]

        # Matplotlib Bar Chart
        plt.figure(figsize=(10, 6))
        plt.bar(categories, counts, color='skyblue')
        plt.xlabel('GRC Category')
        plt.ylabel('Number of Findings')
        plt.title(f'Number of GRC Findings per Category ({analysis_recipe_name})')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        bar_chart_path = os.path.join(output_dir, f'{analysis_recipe_name.lower().replace(" ", "_")}_grc_findings_bar_chart.png')
        plt.savefig(bar_chart_path)
        plt.close()
        visualization_paths.append(bar_chart_path)

        # Plotly Interactive Bar Chart
        hover_text = [f'Category: {cat}<br>Findings: {count}' for cat, count in zip(categories, counts)]
        fig = go.Figure(
            data=[go.Bar(x=categories, y=counts, hoverinfo='text', hovertext=hover_text, marker_color='lightcoral')],
            layout=go.Layout(
                title=go.layout.Title(text=f'Interactive GRC Findings per Category ({analysis_recipe_name})'),
                xaxis=go.layout.XAxis(title_text='GRC Category'),
                yaxis=go.layout.YAxis(title_text='Number of Findings')
            )
        )
        interactive_bar_chart_path = os.path.join(output_dir, f'{analysis_recipe_name.lower().replace(" ", "_")}_grc_findings_interactive_bar_chart.html')
        py.plot(fig, filename=interactive_bar_chart_path, auto_open=False)
        visualization_paths.append(interactive_bar_chart_path)

    # --- Prepare data for Pie Chart (Keyword Distribution) ---
    if extracted_keywords:
        keyword_counts = {kw: extracted_keywords.count(kw) for kw in set(extracted_keywords)}

        labels = list(keyword_counts.keys())
        sizes = list(keyword_counts.values())

        plt.figure(figsize=(10, 7))
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, textprops={'fontsize': 9})
        plt.axis('equal')
        plt.title(f'Distribution of Extracted Keywords ({analysis_recipe_name})')
        plt.tight_layout()
        pie_chart_path = os.path.join(output_dir, f'{analysis_recipe_name.lower().replace(" ", "_")}_grc_keywords_pie_chart.png')
        plt.savefig(pie_chart_path)
        plt.close()
        visualization_paths.append(pie_chart_path)

    # --- Generate GRC Trend Chart (New addition) ---
    trend_chart_path = generate_grc_trend_chart(output_dir)
    if trend_chart_path:
        visualization_paths.append(trend_chart_path)

    print(f"Generated visualizations: {visualization_paths}")
    return visualization_paths

def format_grc_report(grc_output_dict,
                      include_summary=True,
                      include_score_badge=True,
                      include_keywords=True,
                      include_findings=True,
                      include_visualizations=True,
                      output_format='markdown'): # Changed default to markdown
    """
    Formats the raw GRC output dictionary into a human-readable string,
    with configurable sections and output format (text, html, or markdown).

    Args:
        grc_output_dict (dict): The dictionary containing GRC analysis results,
                                 including score and badge.
        include_summary (bool): Whether to include the summary section.
        include_score_badge (bool): Whether to include the GRC score and badge section.
        include_keywords (bool): Whether to include the extracted keywords section.
        include_findings (bool): Whether to include the detailed findings section.
        include_visualizations (bool): Whether to include links/references to visualizations.
        output_format (str): The desired output format ('text', 'html', or 'markdown').

    Returns:
        str: A multi-line string representing the formatted GRC report in the specified format.
    """
    summary = grc_output_dict.get('summary', 'No summary available.')
    findings = grc_output_dict.get('findings', ['No findings to report.'])
    grc_score = grc_output_dict.get('grc_score', 'N/A')
    grc_badge = grc_output_dict.get('grc_badge', 'N/A')
    docling_output = grc_output_dict.get('docling_output', {})
    extracted_keywords = docling_output.get('extracted_keywords', [])
    recipe_used = docling_output.get('recipe_used', 'N/A')
    visualization_paths = grc_output_dict.get('visualization_paths', [])

    if output_format == 'html':
        report_string = """<!DOCTYPE html>
<html>
<head>
<title>GRC Analysis Report</title>
<style>
body { font-family: Arial, sans-serif; line-height: 1.6; color: #333; }
h1 { color: #2c3e50; border-bottom: 2px solid #3498db; padding-bottom: 10px; }
h2 { color: #34495e; }
ul { list-style-type: disc; margin-left: 20px; }
li { margin-bottom: 5px; }
.score-badge { font-weight: bold; color: #e74c3c; }
.gold { color: #ffd700; }
.silver { color: #c0c0c0; }
.bronze { color: #cd7f32; }
img { max-width: 100%; height: auto; margin-top: 10px; border: 1px solid #ddd; }
</style>
</head>
<body>
<h1>GRC Analysis Report</h1>
"""

        if include_summary:
            report_string += f"<h2>Summary</h2><p>{summary}</p>\n"

        if include_score_badge:
            badge_class = grc_badge.lower() if grc_badge != 'N/A' and grc_badge != 'None' else ''
            report_string += f"<h2>GRC Score & Badge</h2><p>GRC Score: <strong>{grc_score}/100</strong><br>GRC Badge: <span class=\"score-badge {badge_class}\">{grc_badge}</span></p>\n"

        report_string += f"<h2>Analysis Details</h2><p>Analysis Recipe Used: <strong>{recipe_used}</strong></p>\n"

        if include_keywords:
            report_string += f"<h3>Extracted Keywords:</h3>\n<ul>\n"
            if extracted_keywords:
                for kw in extracted_keywords:
                    report_string += f"<li>{kw}</li>\n"
            else:
                report_string += "<li>None</li>\n"
            report_string += "</ul>\n"

        if include_findings:
            report_string += f"<h3>Findings:</h3>\n<ul>\n"
            if findings:
                for finding in findings:
                    report_string += f"<li>{finding}</li>\n"
            else:
                report_string += "<li>No findings to report.</li>\n"
            report_string += "</ul>\n"

        if include_visualizations and visualization_paths:
            report_string += "<h3>Visualizations:</h3>\n"
            for viz_path in visualization_paths:
                # Determine if it's an image or an HTML plot
                if viz_path.endswith('.html'):
                    report_string += f"<p><a href='{os.path.basename(viz_path)}' target='_blank'>View Interactive Plot: {os.path.basename(viz_path)}</a></p>\n"
                else:
                    report_string += f"<img src='{os.path.basename(viz_path)}' alt='GRC Visualization'><br>\n"

        report_string += "</body>\n</html>"

    elif output_format == 'markdown': # New markdown format
        report_string = f"""
# GRC Analysis Report
"""
        if include_summary:
            report_string += f"## Summary\n\n{summary}\n\n"

        if include_score_badge:
            report_string += f"## GRC Score & Badge\n\nGRC Score: **{grc_score}/100**\nGRC Badge: **{grc_badge}**\n\n"

        report_string += f"## Analysis Details\n\nAnalysis Recipe Used: **{recipe_used}**\n\n"

        if include_keywords:
            report_string += f"### Extracted Keywords:\n"
            if extracted_keywords:
                for kw in extracted_keywords:
                    report_string += f"- {kw}\n"
            else:
                report_string += "- None\n"
            report_string += "\n"

        if include_findings:
            report_string += f"### Findings:\n"
            if findings:
                for finding in findings:
                    report_string += f"- {finding}\n"
            else:
                report_string += "- No findings to report.\n"
            report_string += "\n"

        if include_visualizations and visualization_paths:
            report_string += "### Visualizations:\n"
            for viz_path in visualization_paths:
                if viz_path.endswith('.html'):
                     report_string += f"- [View Interactive Plot: {os.path.basename(viz_path)}]({os.path.basename(viz_path)})\n"
                else:
                    report_string += f"![GRC Visualization]({os.path.basename(viz_path)})\n\n"

    else:  # Default to 'text' format
        report_string = f"""
--- GRC Analysis Report ---
"""

        if include_summary:
            report_string += f"Summary: {summary}\n\n"

        if include_score_badge:
            report_string += f"GRC Score: {grc_score}/100\n"
            report_string += f"GRC Badge: {grc_badge}\n\n"

        report_string += f"Analysis Recipe Used: {recipe_used}\n"

        if include_keywords:
            report_string += f"Extracted Keywords: {', '.join(extracted_keywords) if extracted_keywords else 'None'}\n\n"

        if include_findings:
            report_string += "Findings:\n"
            for finding in findings:
                report_string += f"- {finding}\n"

        if include_visualizations and visualization_paths:
            report_string += "\nVisualizations:\n"
            for viz_path in visualization_paths:
                report_string += f"- {viz_path}\n"

        report_string += "---------------------------"

    return report_string

# Main GRC document processor function (updated to use format_grc_report with customization options)
def main_grc_doc_processor(document_folder, analysis_recipe_name='Standard_GRC', output_file_path=None, output_format='markdown', output_visualization_dir='grc_visualizations', **report_format_options):
    print(f"Starting GRC document processing in folder: {document_folder} with recipe: {analysis_recipe_name}")
    if not os.path.exists(document_folder):
        print(f"Folder not found: {document_folder}")
        return

    for filename in os.listdir(document_folder):
        file_path = os.path.join(document_folder, filename)
        if os.path.isfile(file_path):
            print(f"\n--- Processing {filename} with recipe {analysis_recipe_name} ---")
            doc_content = read_document(file_path)
            if not doc_content.startswith("Error"):
                docling_output = process_document_for_docling(doc_content, analysis_recipe_name)
                grc_output = perform_grc_checks(docling_output, analysis_recipe_name)
                grc_score = calculate_grc_score(docling_output, analysis_recipe_name) # Calculate the score
                grc_badge = generate_grc_badge(grc_score) # Generate the badge
                grc_output['grc_score'] = grc_score # Add score to the output
                grc_output['grc_badge'] = grc_badge # Add badge to the output

                # Generate visualizations and add their paths to grc_output
                full_output_viz_dir = os.path.join(document_folder, output_visualization_dir)
                visualization_paths = generate_grc_visualizations(grc_output, full_output_viz_dir)
                grc_output['visualization_paths'] = visualization_paths

                # Format the report using the new function with custom options and output format
                formatted_report = format_grc_report(grc_output, output_format=output_format, **report_format_options)

                if output_file_path:
                    base_name, _ = os.path.splitext(output_file_path)
                    if output_format == 'html':
                        final_output_path = f"{base_name}.html"
                    elif output_format == 'markdown': # Handle markdown file extension
                        final_output_path = f"{base_name}.md"
                    else:
                        final_output_path = f"{base_name}.txt"

                    output_dir = os.path.dirname(final_output_path)
                    if output_dir and not os.path.exists(output_dir):
                        os.makedirs(output_dir, exist_ok=True)
                    with open(final_output_path, 'w', encoding='utf-8') as f:
                        f.write(formatted_report)
                    print(f"Final GRC report for {filename} saved to {final_output_path}")
                else:
                    print(f"Final GRC report for {filename}:\n{formatted_report}")
            else:
                print(doc_content)

# Example usage to demonstrate scoring and badge generation
document_directory = 'my_documents'
os.makedirs(document_directory, exist_ok=True)
# Create a dummy file for demonstration, now including 'ethics'
with open(os.path.join(document_directory, 'sample_doc.txt'), 'w') as f:
    f.write("This document discusses compliance, risk management, data protection, privacy, and cybersecurity. It also mentions ethics.")

# --- Demonstrating 'Standard_GRC' recipe with full formatted HTML output including visualizations ---
print("\n--- Demonstrating 'Standard_GRC' recipe with full formatted HTML output including visualizations ---")
output_html_file_with_viz = os.path.join(document_directory, 'standard_grc_report_with_viz.html')
main_grc_doc_processor(document_directory, analysis_recipe_name='Standard_GRC', output_file_path=output_html_file_with_viz, output_format='html', include_visualizations=True)

print(f"\nContent of {output_html_file_with_viz}: (showing first 1000 chars)\n")
if os.path.exists(output_html_file_with_viz):
    with open(output_html_file_with_viz, 'r', encoding='utf-8') as f:
        print(f.read()[:1000])
else:
    print(f"File not found: {output_html_file_with_viz}")

# --- Demonstrating 'Privacy_Focused_GRC' recipe with full formatted HTML output including visualizations ---
print("\n--- Demonstrating 'Privacy_Focused_GRC' recipe with full formatted HTML output including visualizations ---")
output_html_file_with_viz_privacy = os.path.join(document_directory, 'privacy_grc_report_with_viz.html')
main_grc_doc_processor(document_directory, analysis_recipe_name='Privacy_Focused_GRC', output_file_path=output_html_file_with_viz_privacy, output_format='html', include_visualizations=True)

print(f"\nContent of {output_html_file_with_viz_privacy}: (showing first 1000 chars)\n")
if os.path.exists(output_html_file_with_viz_privacy):
    with open(output_html_file_with_viz_privacy, 'r', encoding='utf-8') as f:
        print(f.read()[:1000])
else:
    print(f"File not found: {output_html_file_with_viz_privacy}")

# --- Demonstrating 'Standard_GRC' recipe with full formatted Markdown output including visualizations ---
print("\n--- Demonstrating 'Standard_GRC' recipe with full formatted Markdown output including visualizations ---")
output_md_file_with_viz = os.path.join(document_directory, 'standard_grc_report_with_viz.md')
main_grc_doc_processor(document_directory, analysis_recipe_name='Standard_GRC', output_file_path=output_md_file_with_viz, output_format='markdown', include_visualizations=True)

print(f"\nContent of {output_md_file_with_viz}: (showing first 1000 chars)\n")
if os.path.exists(output_md_file_with_viz):
    with open(output_md_file_with_viz, 'r', encoding='utf-8') as f:
        print(f.read()[:1000])
else:
    print(f"File not found: {output_md_file_with_viz}")
```

```python
import os
import PyPDF2 # For PDF files
import docx # For DOCX files
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.offline as py
from datetime import datetime, timedelta

# --- ANALYSIS_RECIPES: Consolidated from all previous modifications ---
ANALYSIS_RECIPES = {
    'Standard_GRC': {
        'keywords': ['compliance', 'regulation', 'risk', 'audit', 'policy', 'governance', 'security', 'privacy', 'ethics'],
        'grc_categories': {
            'Compliance': ['compliance', 'regulation', 'policy'],
            'Risk Management': ['risk', 'security', 'privacy'],
            'Governance': ['governance', 'audit'],
            'Ethics': ['ethics']
        },
        'category_weights': {
            'Compliance': 5,
            'Risk Management': 10,
            'Governance': 3,
            'Ethics': 7
        }
    },
    'Privacy_Focused_GRC': {
        'keywords': ['privacy', 'data protection', 'GDPR', 'CCPA', 'PII', 'consent', 'security', 'compliance'],
        'grc_categories': {
            'Data Privacy': ['privacy', 'data protection', 'GDPR', 'CCPA', 'PII', 'consent'],
            'Security Compliance': ['security', 'compliance']
        },
        'category_weights': {
            'Data Privacy': 8,
            'Security Compliance': 6
        }
    },
    'Security_Focused_GRC': {
        'keywords': ['security', 'cybersecurity', 'breach', 'vulnerability', 'encryption', 'firewall', 'incident', 'data loss'],
        'grc_categories': {
            'Data Security': ['security', 'cybersecurity', 'encryption', 'firewall'],
            'Incident Response': ['breach', 'vulnerability', 'incident', 'data loss']
        },
        'category_weights': {
            'Data Security': 7,
            'Incident Response': 10
        }
    },
    'Legal_Compliance_GRC': {
        'keywords': ['legal', 'law', 'regulation', 'contract', 'litigation', 'fine', 'penalty', 'jurisdiction'],
        'grc_categories': {
            'Legal Compliance': ['legal', 'law', 'regulation', 'contract'],
            'Legal Risk': ['litigation', 'fine', 'penalty', 'jurisdiction']
        },
        'category_weights': {
            'Legal Compliance': 6,
            'Legal Risk': 12
        }
    },
    'OEI_Organisational_Ethics_Indicator': {
        'keywords': ['ethics', 'integrity', 'values', 'conduct', 'culture', 'governance', 'transparency', 'accountability', 'compliance', 'misconduct', 'whistleblower', 'dignity', 'respect', 'fairness'],
        'grc_categories': {
            'Ethical Culture': ['ethics', 'integrity', 'values', 'culture', 'respect', 'dignity', 'fairness'],
            'Conduct and Accountability': ['conduct', 'transparency', 'accountability', 'governance', 'misconduct'],
            'Whistleblower Protection': ['whistleblower', 'reporting', 'retaliation'],
            'Ethics Compliance': ['compliance', 'code of conduct']
        },
        'category_weights': {
            'Ethical Culture': 7,
            'Conduct and Accountability': 9,
            'Whistleblower Protection': 10,
            'Ethics Compliance': 8
        }
    },
    'Framework_Gap_Analysis_GRC': {
        'keywords': ['framework', 'standard', 'compliance', 'gap', 'assessment', 'alignment', 'recommendation', 'control', 'policy', 'procedure', 'best practice', 'regulatory', 'deficiency'],
        'grc_categories': {
            'Framework Alignment': ['framework', 'standard', 'alignment', 'best practice'],
            'Gap Identification': ['gap', 'assessment', 'deficiency'],
            'Remediation Planning': ['recommendation', 'control', 'policy', 'procedure'],
            'Regulatory Compliance': ['compliance', 'regulatory']
        },
        'category_weights': {
            'Framework Alignment': 5,
            'Gap Identification': 9,
            'Remediation Planning': 7,
            'Regulatory Compliance': 8
        }
    },
    'Vendor_Risk_Management_GRC': {
        'keywords': ['vendor', 'third-party', 'contract', 'SLA', 'service level', 'due diligence', 'onboarding', 'offboarding', 'data sharing', 'supply chain', 'resilience', 'security breach'],
        'grc_categories': {
            'Vendor Due Diligence': ['vendor', 'third-party', 'due diligence', 'onboarding'],
            'Contractual Compliance': ['contract', 'SLA', 'service level', 'data sharing'],
            'Supply Chain Risk': ['supply chain', 'resilience', 'security breach']
        },
        'category_weights': {
            'Vendor Due Diligence': 6,
            'Contractual Compliance': 8,
            'Supply Chain Risk': 11
        }
    },
    'ESG_Sustainability_GRC': {
        'keywords': ['sustainability', 'environmental', 'social', 'governance', 'carbon footprint', 'emissions', 'diversity', 'inclusion', 'community', 'ethics', 'CSR', 'reporting', 'impact', 'stakeholder'],
        'grc_categories': {
            'Environmental Impact': ['sustainability', 'environmental', 'carbon footprint', 'emissions'],
            'Social Responsibility': ['social', 'diversity', 'inclusion', 'community', 'CSR', 'stakeholder'],
            'Governance & Ethics (ESG)': ['governance', 'ethics', 'reporting', 'impact']
        },
        'category_weights': {
            'Environmental Impact': 7,
            'Social Responsibility': 6,
            'Governance & Ethics (ESG)': 8
        }
    },
    'Corporate_Governance_Compliance_GRC': {
        'keywords': ['corporate governance', 'board of directors', 'shareholder rights', 'executive compensation', 'internal controls', 'regulatory reporting', 'fiduciary duty', 'disclosure', 'ethics committee'],
        'grc_categories': {
            'Board Oversight': ['board of directors', 'fiduciary duty', 'governance'],
            'Executive Accountability': ['executive compensation', 'executive conduct'],
            'Transparency & Disclosure': ['shareholder rights', 'regulatory reporting', 'disclosure'],
            'Internal Controls & Ethics': ['internal controls', 'ethics committee']
        },
        'category_weights': {
            'Board Oversight': 9,
            'Executive Accountability': 8,
            'Transparency & Disclosure': 7,
            'Internal Controls & Ethics': 6
        }
    },
    'Reputational_Risk_GRC': {
        'keywords': ['reputation', 'brand image', 'public trust', 'stakeholder perception', 'media coverage', 'ethical lapse', 'controversy', 'scandal', 'integrity', 'social media', 'consumer sentiment'],
        'grc_categories': {
            'Public Perception & Trust': ['reputation', 'brand image', 'public trust', 'stakeholder perception', 'consumer sentiment'],
            'Media & Communication Risk': ['media coverage', 'social media', 'controversy'],
            'Ethical Conduct Impact': ['ethical lapse', 'scandal', 'integrity']
        },
        'category_weights': {
            'Public Perception & Trust': 10,
            'Media & Communication Risk': 9,
            'Ethical Conduct Impact': 12
        }
    },
    'Legal_Jurisprudence_GRC': {
        'keywords': [
            'jurisprudence', 'precedent', 'common law', 'statutory law', 'case law',
            'constitutional law', 'judicial review', 'legal theory', 'court decision', 'legal interpretation'
        ],
        'grc_categories': {
            'Legal Theory': ['jurisprudence', 'legal theory', 'legal interpretation'],
            'Judicial Precedent': ['precedent', 'common law', 'case law', 'court decision'],
            'Constitutional Aspects': ['constitutional law', 'judicial review', 'statutory law']
        },
        'category_weights': {
            'Legal Theory': 7,
            'Judicial Precedent': 9,
            'Constitutional Aspects': 10
        }
    },
    'Cultural_Context_GRC': {
        'keywords': [
            'culture', 'organizational values', 'ethics', 'diversity', 'inclusion',
            'employee engagement', 'leadership', 'communication', 'trust',
            'behavior', 'norms', 'tradition', 'local custom', 'social fabric'
        ],
        'grc_categories': {
            'Organizational Culture': ['culture', 'organizational values', 'employee engagement', 'leadership', 'trust'],
            'Ethical & Social Norms': ['ethics', 'diversity', 'inclusion', 'behavior', 'norms'],
            'Local & Traditional Impact': ['tradition', 'local custom', 'social fabric', 'communication']
        },
        'category_weights': {
            'Organizational Culture': 8,
            'Ethical & Social Norms': 9,
            'Local & Traditional Impact': 7
        }
    },
    'Economic_Impact_GRC': {
        'keywords': [
            'economic impact', 'financial risk', 'market volatility', 'revenue', 'profit',
            'investment', 'shareholder value', 'cost reduction', 'sustainability', 'growth',
            'inflation', 'recession', 'supply chain disruption'
        ],
        'grc_categories': {
            'Financial Performance': ['economic impact', 'revenue', 'profit', 'shareholder value', 'growth'],
            'Market & Investment Risk': ['financial risk', 'market volatility', 'investment', 'inflation', 'recession'],
            'Operational Economics': ['cost reduction', 'supply chain disruption']
        },
        'category_weights': {
            'Financial Performance': 9,
            'Market & Investment Risk': 11,
            'Operational Economics': 8
        }
    },
    'Social_Impact_GRC': {
        'keywords': [
            'social impact', 'community engagement', 'human rights', 'labor practices',
            'diversity', 'inclusion', 'stakeholder relations', 'philanthropy',
            'public health', 'education access', 'social equity', 'ethical sourcing'
        ],
        'grc_categories': {
            'Community & Stakeholder Impact': ['social impact', 'community engagement', 'stakeholder relations', 'philanthropy'],
            'Human Capital & Rights': ['human rights', 'labor practices', 'diversity', 'inclusion', 'ethical sourcing'],
            'Societal Welfare': ['public health', 'education access', 'social equity']
        },
        'category_weights': {
            'Community & Stakeholder Impact': 8,
            'Human Capital & Rights': 10,
            'Societal Welfare': 7
        }
    },
    'Company_Profile_GRC': {
        'keywords': [
            'company overview', 'mission statement', 'vision', 'corporate values', 'organizational structure',
            'leadership team', 'board of directors', 'annual report', 'financial performance', 'market position',
            'industry', 'business model', 'stakeholders', 'CSR initiatives', 'sustainability goals'
        ],
        'grc_categories': {
            'Corporate Identity': ['company overview', 'mission statement', 'vision', 'corporate values'],
            'Organizational Governance': ['organizational structure', 'leadership team', 'board of directors'],
            'Business Context': ['annual report', 'financial performance', 'market position', 'industry', 'business model'],
            'Social & Environmental Stance': ['stakeholders', 'CSR initiatives', 'sustainability goals']
        },
        'category_weights': {
            'Corporate Identity': 5,
            'Organizational Governance': 7,
            'Business Context': 6,
            'Social & Environmental Stance': 8
        }
    },
    'Country_Jurisdictional_GRC': {
        'keywords': [
            'jurisdiction', 'country law', 'national regulation', 'local statutes', 'international treaties',
            'taxation policy', 'labor law', 'environmental regulations', 'data privacy laws', 'sanctions',
            'import export rules', 'trade agreements', 'political stability', 'economic climate', 'cultural norms'
        ],
        'grc_categories': {
            'Legal & Regulatory Framework': ['jurisdiction', 'country law', 'national regulation', 'local statutes', 'international treaties'],
            'Economic & Trade Compliance': ['taxation policy', 'import export rules', 'trade agreements', 'sanctions'],
            'Socio-Political Landscape': ['labor law', 'environmental regulations', 'data privacy laws', 'political stability', 'economic climate', 'cultural norms']
        },
        'category_weights': {
            'Legal & Regulatory Framework': 12,
            'Economic & Trade Compliance': 10,
            'Socio-Political Landscape': 8
        }
    },
    'KPI_Performance_GRC': {
        'keywords': [
            'KPI', 'key performance indicator', 'metric', 'performance target', 'reporting period', 'achievement',
            'underperformance', 'goal', 'objective', 'benchmark', 'audit findings', 'compliance rate', 'risk exposure',
            'governance effectiveness', 'ethics score'
        ],
        'grc_categories': {
            'Performance Measurement': ['KPI', 'key performance indicator', 'metric', 'performance target', 'goal', 'objective', 'benchmark'],
            'Compliance & Risk Performance': ['audit findings', 'compliance rate', 'risk exposure'],
            'Governance & Ethics Performance': ['governance effectiveness', 'ethics score', 'achievement', 'underperformance']
        },
        'category_weights': {
            'Performance Measurement': 9,
            'Compliance & Risk Performance': 11,
            'Governance & Ethics Performance': 8
        }
    }
}


# --- REPORT_SECTION_RECIPES: Consolidated from previous modifications ---
REPORT_SECTION_RECIPES = {
    'Full_GRC_Report_Template': [
        {
            'name': 'Report Title',
            'title': 'GRC Analysis Report',
            'type': 'static_text',
            'content': None,
            'display_order': 1
        },
        {
            'name': 'Summary',
            'title': 'Executive Summary',
            'type': 'dynamic_text',
            'content_key': 'summary',
            'display_order': 2
        },
        {
            'name': 'Score and Badge',
            'title': 'GRC Score & Badge',
            'type': 'dynamic_score_badge',
            'content_key': None,
            'display_order': 3
        },
        {
            'name': 'Analysis Details',
            'title': 'Analysis Details',
            'type': 'dynamic_text',
            'content_key': 'docling_output.recipe_used',
            'format_options': {'prefix': 'Analysis Recipe Used: '},
            'display_order': 4
        },
        {
            'name': 'Extracted Keywords',
            'title': 'Extracted Keywords',
            'type': 'dynamic_list',
            'content_key': 'docling_output.extracted_keywords',
            'format_options': {'list_type': 'unordered'},
            'display_order': 5
        },
        {
            'name': 'Findings',
            'title': 'Findings',
            'type': 'dynamic_list',
            'content_key': 'findings',
            'format_options': {'list_type': 'unordered'},
            'display_order': 6
        },
        {
            'name': 'Visualizations',
            'title': 'Visualizations',
            'type': 'dynamic_visualizations',
            'content_key': 'visualization_paths',
            'display_order': 7
        },
        {
            'name': 'Recommendations',
            'title': 'Recommendations',
            'type': 'static_text',
            'content': 'Based on the analysis, consider reviewing identified high-risk areas and implementing robust control measures.',
            'display_order': 8
        },
        {
            'name': 'Conclusion',
            'title': 'Conclusion',
            'type': 'static_text',
            'content': 'This report provides a comprehensive overview of the document\'s GRC posture.',
            'display_order': 9
        }
    ],
    'Executive_Summary_Template': [
        {
            'name': 'Report Title',
            'title': 'Executive GRC Summary',
            'type': 'static_text',
            'content': None,
            'display_order': 1
        },
        {
            'name': 'Summary',
            'title': 'Overview',
            'type': 'dynamic_text',
            'content_key': 'summary',
            'display_order': 2
        },
        {
            'name': 'Score and Badge',
            'title': 'Overall GRC Rating',
            'type': 'dynamic_score_badge',
            'content_key': None,
            'display_order': 3
        }
    ],
    'Findings_Only_Template': [
        {
            'name': 'Report Title',
            'title': 'Detailed GRC Findings Report',
            'type': 'static_text',
            'content': None,
            'display_order': 1
        },
        {
            'name': 'Findings',
            'title': 'Key Findings',
            'type': 'dynamic_list',
            'content_key': 'findings',
            'format_options': {'list_type': 'unordered'},
            'display_order': 2
        }
    ]
}


def _get_nested_value(data, key_path):
    """
    Retrieves a value from a nested dictionary using a dot-separated key path.
    """
    keys = key_path.split('.')
    current_value = data
    for key in keys:
        if isinstance(current_value, dict) and key in current_value:
            current_value = current_value[key]
        else:
            return None # Key not found or path invalid
    return current_value

# Function to simulate document parsing/reading
def read_document(file_path):
    print(f"Reading document: {file_path}")
    try:
        if file_path.lower().endswith('.pdf'):
            with open(file_path, 'rb') as f:
                reader = PyPDF2.PdfReader(f)
                content = "".join([page.extract_text() for page in reader.pages if page.extract_text()])
            return content
        elif file_path.lower().endswith('.docx'):
            document = docx.Document(file_path)
            content = "\n".join([para.text for para in document.paragraphs])
            return content
        else:
            # Assume plain text for other file types
            with open(file_path, 'r', encoding='utf-8') as f:
                content = f.read()
            return content
    except FileNotFoundError:
        return f"Error: Document not found at {file_path}"
    except Exception as e:
        return f"Error reading document: {e}"

# Function to simulate 'docling' processing
def process_document_for_docling(document_content, analysis_recipe_name='Standard_GRC'):
    print(f"Processing document content for 'docling' using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for keyword extraction.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    keywords = recipe_to_use['keywords']
    found_keywords = [kw for kw in keywords if kw in document_content.lower()]

    processed_data = {
        'document_summary': document_content[:100],
        'extracted_keywords': found_keywords,
        'keyword_count': len(found_keywords),
        'recipe_used': analysis_recipe_name
    }
    print(f"Extracted keywords: {found_keywords}")
    return processed_data

# Function for GRC related operations
def perform_grc_checks(processed_data, analysis_recipe_name='Standard_GRC'):
    print(f"Performing GRC checks using recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found. Using 'Standard_GRC' for GRC checks.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    grc_categories = recipe_to_use['grc_categories']
    extracted_keywords = processed_data.get('extracted_keywords', [])

    grc_findings = []
    categorized_findings = {}
    for category, category_keywords in grc_categories.items():
        for kw in category_keywords:
            if kw in extracted_keywords:
                if category not in categorized_findings:
                    categorized_findings[category] = []
                categorized_findings[category].append(f"Keyword '{kw}' detected in {category} area.")

    if categorized_findings:
        for category, findings in categorized_findings.items():
            grc_findings.append(f"{category} Findings: {'; '.join(findings)}")
    else:
        grc_findings.append(f"No specific GRC flags based on defined categories for {analysis_recipe_name}.")

    grc_report = {
        'summary': f"GRC analysis based on '{analysis_recipe_name}' recipe output.",
        'findings': grc_findings,
        'docling_output': processed_data,
        'recipe_used_for_checks': analysis_recipe_name
    }
    return grc_report

def calculate_grc_score(processed_data, analysis_recipe_name='Standard_GRC'):
    max_score = 100
    current_score = max_score
    default_deduction = 5 # Default weight for categories not explicitly listed

    print(f"Calculating GRC score for recipe: {analysis_recipe_name}...")

    if analysis_recipe_name not in ANALYSIS_RECIPES:
        print(f"Warning: Recipe '{analysis_recipe_name}' not found for scoring. Using 'Standard_GRC'.")
        recipe_to_use = ANALYSIS_RECIPES['Standard_GRC']
    else:
        recipe_to_use = ANALYSIS_RECIPES[analysis_recipe_name]

    grc_categories = recipe_to_use.get('grc_categories', {})
    category_weights = recipe_to_use.get('category_weights', {})
    extracted_keywords = processed_data.get('extracted_keywords', [])

    findings_categories = set()
    for category, category_keywords in grc_categories.items():
        for kw in category_keywords:
            if kw in extracted_keywords:
                findings_categories.add(category)
                break # Only need to find one keyword to mark category as having findings

    for category_with_finding in findings_categories:
        deduction = category_weights.get(category_with_finding, default_deduction)
        print(f"  - Found findings in category '{category_with_finding}', deducting {deduction} points.")
        current_score -= deduction
        if current_score < 0:
            current_score = 0
            break

    print(f"GRC Score: {current_score}/{max_score}")
    return current_score

def generate_grc_badge(grc_score):
    """
    Generates a textual representation of a GRC badge based on the GRC score.

    Args:
        grc_score (int): The calculated GRC score.

    Returns:
        str: The badge level (e.g., 'Gold', 'Silver', 'Bronze', 'None').
    """
    if grc_score >= 90:
        return 'Gold'
    elif grc_score >= 70:
        return 'Silver'
    elif grc_score >= 50:
        return 'Bronze'
    else:
        return 'None'

def generate_grc_trend_chart(output_dir):
    """
    Generates a trend line chart of conceptual GRC scores over time.

    Args:
        output_dir (str): The directory to save the generated plot.

    Returns:
        str: The file path to the saved trend line chart image, or None if not generated.
    """
    os.makedirs(output_dir, exist_ok=True)

    # Simulate historical GRC score data (e.g., last 6 months)
    end_date = datetime.now()
    dates = [end_date - timedelta(days=30 * i) for i in range(6, -1, -1)] # 7 data points for 6 months
    # Dummy scores, with a conceptual 'trend' or fluctuation
    scores = [70, 75, 80, 78, 85, 82, 88]

    if not dates or not scores:
        print("No data available to generate GRC trend chart.")
        return None

    plt.figure(figsize=(10, 6))
    plt.plot(dates, scores, marker='o', linestyle='-', color='green')
    plt.xlabel('Date')
    plt.ylabel('GRC Score')
    plt.title('Conceptual GRC Score Trend Over Time')
    plt.grid(True)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    trend_chart_path = os.path.join(output_dir, 'grc_score_trend_chart.png')
    plt.savefig(trend_chart_path)
    plt.close()
    print(f"Generated GRC trend chart: {trend_chart_path}")
    return trend_chart_path

def generate_grc_visualizations(grc_output_dict, output_dir):
    """
    Generates visual representations (bar chart, pie chart, interactive bar chart, trend chart) from GRC output data.

    Args:
        grc_output_dict (dict): The dictionary containing GRC analysis results.
        output_dir (str): The directory to save the generated plots.

    Returns:
        list: A list of file paths to the saved visualization images/html files.
    """
    os.makedirs(output_dir, exist_ok=True)
    visualization_paths = []

    # Extract data for visualizations
    findings = grc_output_dict.get('findings', [])
    extracted_keywords = grc_output_dict.get('docling_output', {}).get('extracted_keywords', [])
    analysis_recipe_name = grc_output_dict.get('docling_output', {}).get('recipe_used', 'N/A')

    # --- Prepare data for Bar Chart (Findings per Category) ---
    findings_counts = {}
    for finding_str in findings:
        if ':' in finding_str:
            category_name = finding_str.split(':')[0].strip()
            if ' Findings' in category_name:
                category_name = category_name.replace(' Findings', '')
            findings_counts[category_name] = findings_counts.get(category_name, 0) + 1

    # Sort findings_counts by category name for consistent plotting
    sorted_findings = sorted(findings_counts.items())

    if sorted_findings:
        categories = [item[0] for item in sorted_findings]
        counts = [item[1] for item in sorted_findings]

        # Matplotlib Bar Chart
        plt.figure(figsize=(10, 6))
        plt.bar(categories, counts, color='skyblue')
        plt.xlabel('GRC Category')
        plt.ylabel('Number of Findings')
        plt.title(f'Number of GRC Findings per Category ({analysis_recipe_name})')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        bar_chart_path = os.path.join(output_dir, f'{analysis_recipe_name.lower().replace(" ", "_")}_grc_findings_bar_chart.png')
        plt.savefig(bar_chart_path)
        plt.close()
        visualization_paths.append(bar_chart_path)

        # Plotly Interactive Bar Chart
        hover_text = [f'Category: {cat}<br>Findings: {count}' for cat, count in zip(categories, counts)]
        fig = go.Figure(
            data=[go.Bar(x=categories, y=counts, hoverinfo='text', hovertext=hover_text, marker_color='lightcoral')],
            layout=go.Layout(
                title=go.layout.Title(text=f'Interactive GRC Findings per Category ({analysis_recipe_name})'),
                xaxis=go.layout.XAxis(title_text='GRC Category'),
                yaxis=go.layout.YAxis(title_text='Number of Findings')
            )
        )
        interactive_bar_chart_path = os.path.join(output_dir, f'{analysis_recipe_name.lower().replace(" ", "_")}_grc_findings_interactive_bar_chart.html')
        py.plot(fig, filename=interactive_bar_chart_path, auto_open=False)
        visualization_paths.append(interactive_bar_chart_path)

    # --- Prepare data for Pie Chart (Keyword Distribution) ---
    if extracted_keywords:
        keyword_counts = {kw: extracted_keywords.count(kw) for kw in set(extracted_keywords)}

        labels = list(keyword_counts.keys())
        sizes = list(keyword_counts.values())

        plt.figure(figsize=(10, 7))
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, textprops={'fontsize': 9})
        plt.axis('equal')
        plt.title(f'Distribution of Extracted Keywords ({analysis_recipe_name})')
        plt.tight_layout()
        pie_chart_path = os.path.join(output_dir, f'{analysis_recipe_name.lower().replace(" ", "_")}_grc_keywords_pie_chart.png')
        plt.savefig(pie_chart_path)
        plt.close()
        visualization_paths.append(pie_chart_path)

    # --- Generate GRC Trend Chart (New addition) ---
    trend_chart_path = generate_grc_trend_chart(output_dir)
    if trend_chart_path:
        visualization_paths.append(trend_chart_path)

    print(f"Generated visualizations: {visualization_paths}")
    return visualization_paths

def format_grc_report(grc_output_dict, report_recipe_name='Full_GRC_Report_Template', output_format='markdown'):
    """
    Formats the raw GRC output dictionary into a human-readable string based on a report recipe.

    Args:
        grc_output_dict (dict): The dictionary containing GRC analysis results.
        report_recipe_name (str): The name of the report recipe to use from REPORT_SECTION_RECIPES.
        output_format (str): The desired output format ('text', 'html', or 'markdown').

    Returns:
        str: A multi-line string representing the formatted GRC report in the specified format.
    """
    if report_recipe_name not in REPORT_SECTION_RECIPES:
        print(f"Warning: Report recipe '{report_recipe_name}' not found. Using 'Full_GRC_Report_Template'.")
        report_recipe_name = 'Full_GRC_Report_Template'

    report_sections = sorted(REPORT_SECTION_RECIPES[report_recipe_name], key=lambda x: x.get('display_order', 99))

    report_content_parts = []

    # HTML header and styling
    if output_format == 'html':
        report_content_parts.append("""<!DOCTYPE html>
<html>
<head>
<title>GRC Analysis Report</title>
<style>
body { font-family: Arial, sans-serif; line-height: 1.6; color: #333; }
h1 { color: #2c3e50; border-bottom: 2px solid #3498db; padding-bottom: 10px; }
h2 { color: #34495e; }
h3 { color: #5a7d9a; }
ul { list-style-type: disc; margin-left: 20px; }
li { margin-bottom: 5px; }
.score-badge { font-weight: bold; color: #e74c3c; }
.gold { color: #ffd700; }
.silver { color: #c0c0c0; }
.bronze { color: #cd7f32; }
img { max-width: 100%; height: auto; margin-top: 10px; border: 1px solid #ddd; }
</style>
</head>
<body>
""")

    # Main report title (usually the first section)
    main_title_section = next((s for s in report_sections if s['name'] == 'Report Title'), None)
    if main_title_section:
        title_text = main_title_section.get('title', 'GRC Report')
        if output_format == 'html':
            report_content_parts.append(f"<h1>{title_text}</h1>\n")
        elif output_format == 'markdown':
            report_content_parts.append(f"# {title_text}\n")
        else:
            report_content_parts.append(f"--- {title_text} ---\n")

    for section in report_sections:
        if section['name'] == 'Report Title': # Already handled above
            continue

        section_title = section.get('title', section['name'])
        section_type = section['type']
        content_key = section.get('content_key')
        format_options = section.get('format_options', {})

        section_output = ""

        if section_type == 'static_text':
            content = section.get('content', '')
            if content:
                if output_format == 'html':
                    section_output += f"<h2>{section_title}</h2><p>{content}</p>\n"
                elif output_format == 'markdown':
                    section_output += f"## {section_title}\n\n{content}\n\n"
                else:
                    section_output += f"\n--- {section_title} ---\n{content}\n"

        elif section_type == 'dynamic_text':
            content = _get_nested_value(grc_output_dict, content_key)
            prefix = format_options.get('prefix', '')
            if content:
                if output_format == 'html':
                    section_output += f"<h2>{section_title}</h2><p>{prefix}<strong>{content}</strong></p>\n"
                elif output_format == 'markdown':
                    section_output += f"## {section_title}\n\n{prefix}**{content}**\n\n"
                else:
                    section_output += f"\n--- {section_title} ---\n{prefix}{content}\n"
            elif prefix and output_format != 'html' and output_format != 'markdown': # for text output, show prefix even if content is None
                 section_output += f"\n--- {section_title} ---\n{prefix}None\n"

        elif section_type == 'dynamic_score_badge':
            grc_score = grc_output_dict.get('grc_score', 'N/A')
            grc_badge = grc_output_dict.get('grc_badge', 'N/A')
            if output_format == 'html':
                badge_class = grc_badge.lower() if grc_badge != 'N/A' and grc_badge != 'None' else ''
                section_output += f"<h2>{section_title}</h2><p>GRC Score: <strong>{grc_score}/100</strong><br>GRC Badge: <span class=\"score-badge {badge_class}\">{grc_badge}</span></p>\n"
            elif output_format == 'markdown':
                section_output += f"## {section_title}\n\nGRC Score: **{grc_score}/100**\nGRC Badge: **{grc_badge}**\n\n"
            else:
                section_output += f"\n--- {section_title} ---\nGRC Score: {grc_score}/100\nGRC Badge: {grc_badge}\n"

        elif section_type == 'dynamic_list':
            items = _get_nested_value(grc_output_dict, content_key)
            if items:
                if output_format == 'html':
                    section_output += f"<h3>{section_title}:</h3>\n<ul>\n"
                    for item in items:
                        section_output += f"<li>{item}</li>\n"
                    section_output += "</ul>\n"
                elif output_format == 'markdown':
                    section_output += f"### {section_title}:\n"
                    for item in items:
                        section_output += f"- {item}\n"
                    section_output += "\n"
                else:
                    section_output += f"\n--- {section_title} ---\n"
                    for item in items:
                        section_output += f"- {item}\n"
            else:
                if output_format == 'html':
                    section_output += f"<h3>{section_title}:</h3>\n<ul><li>None</li></ul>\n"
                elif output_format == 'markdown':
                    section_output += f"### {section_title}:\n- None\n\n"
                else:
                    section_output += f"\n--- {section_title} ---\nNone\n"

        elif section_type == 'dynamic_visualizations':
            visualization_paths = grc_output_dict.get('visualization_paths', [])
            if visualization_paths:
                if output_format == 'html':
                    section_output += f"<h3>{section_title}:</h3>\n"
                    for viz_path in visualization_paths:
                        if viz_path.endswith('.html'):
                            section_output += f"<p><a href='{os.path.basename(viz_path)}' target='_blank'>View Interactive Plot: {os.path.basename(viz_path)}</a></p>\n"
                        else:
                            section_output += f"<img src='{os.path.basename(viz_path)}' alt='GRC Visualization'><br>\n"
                elif output_format == 'markdown':
                    section_output += f"### {section_title}:\n"
                    for viz_path in visualization_paths:
                        if viz_path.endswith('.html'):
                             report_string += f"- [View Interactive Plot: {os.path.basename(viz_path)}]({os.path.basename(viz_path)})\n"
                        else:
                            section_output += f"![GRC Visualization]({os.path.basename(viz_path)})\n\n"
                else:
                    section_output += f"\n--- {section_title} ---\n"
                    for viz_path in visualization_paths:
                        section_output += f"- {viz_path}\n"
            else:
                if output_format == 'html':
                    section_output += f"<h3>{section_title}:</h3><p>No visualizations available.</p>\n"
                elif output_format == 'markdown':
                    section_output += f"### {section_title}:\nNo visualizations available.\n\n"
                else:
                    section_output += f"\n--- {section_title} ---\nNo visualizations available.\n"
        report_content_parts.append(section_output)

    if output_format == 'html':
        report_content_parts.append("</body>\n</html>")
    elif output_format == 'text':
        report_content_parts.append("---------------------------")

    return "".join(report_content_parts)

# Main GRC document processor function (updated to use format_grc_report with report_recipe_name)
def main_grc_doc_processor(document_folder, analysis_recipe_name='Standard_GRC', output_file_path=None, report_recipe_name='Full_GRC_Report_Template', output_format='markdown', output_visualization_dir='grc_visualizations'):
    print(f"Starting GRC document processing in folder: {document_folder} with recipe: {analysis_recipe_name}")
    if not os.path.exists(document_folder):
        print(f"Folder not found: {document_folder}")
        return

    for filename in os.listdir(document_folder):
        file_path = os.path.join(document_folder, filename)
        if os.path.isfile(file_path):
            print(f"\n--- Processing {filename} with recipe {analysis_recipe_name} ---")
            doc_content = read_document(file_path)
            if not doc_content.startswith("Error"):
                docling_output = process_document_for_docling(doc_content, analysis_recipe_name)
                grc_output = perform_grc_checks(docling_output, analysis_recipe_name)
                grc_score = calculate_grc_score(docling_output, analysis_recipe_name) # Calculate the score
                grc_badge = generate_grc_badge(grc_score) # Generate the badge
                grc_output['grc_score'] = grc_score # Add score to the output
                grc_output['grc_badge'] = grc_badge # Add badge to the output

                # Generate visualizations and add their paths to grc_output
                full_output_viz_dir = os.path.join(document_folder, output_visualization_dir)
                visualization_paths = generate_grc_visualizations(grc_output, full_output_viz_dir)
                grc_output['visualization_paths'] = visualization_paths

                # Format the report using the new function with custom options and output format
                formatted_report = format_grc_report(grc_output, report_recipe_name=report_recipe_name, output_format=output_format)

                if output_file_path:
                    base_name, _ = os.path.splitext(output_file_path)
                    if output_format == 'html':
                        final_output_path = f"{base_name}.html"
                    elif output_format == 'markdown': # Handle markdown file extension
                        final_output_path = f"{base_name}.md"
                    else:
                        final_output_path = f"{base_name}.txt"

                    output_dir = os.path.dirname(final_output_path)
                    if output_dir and not os.path.exists(output_dir):
                        os.makedirs(output_dir, exist_ok=True)
                    with open(final_output_path, 'w', encoding='utf-8') as f:
                        f.write(formatted_report)
                    print(f"Final GRC report for {filename} saved to {final_output_path}")
                else:
                    print(f"Final GRC report for {filename}:\n{formatted_report}")
            else:
                print(doc_content)

print("All functions (including `format_grc_report` and `main_grc_doc_processor`) have been updated to support report section recipes and dynamic content generation.")
```
